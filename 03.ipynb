## 让回归更易懂

## 双校记

- 在2012—2013学年,进入美国四年制私立大学的学生平均每年要支付大约29000美元的学费。那些进入自己所在州公立大学的学生,支付的学费则不到9000美元。精英式私立教育可能在很多方面更好:班级规模更小、体育设施更新、教师更加杰出,学生也更加聪明。但是,每年学费高出20000美元也令人咋舌。你也许会好奇,是否值得花这么多学费去上私立大学。

- 在这个例子中,可比性问题是,在马萨诸塞州出生的,已经年近40岁的哈佛大学毕业生,如果他当年上的是马萨诸塞州立大学(University
  of
  Massachusetts,U-Mass),他现在会赚多少钱。因此,当我们提出为私立学校支付的额外学费是否值得这个问题时,我们关注的是那些进入精英私立大学的人士毕业后可能获得的收入。

- 比较不同类型学生的学生之间的收入差距,必然得出精英大学的校友们收入奇高的结论。然而,略加思考,就可以看出为什么比较哈佛大学和马萨诸塞州立大学的毕业生收入,并不能揭示哈佛大学学位所带来的收益。这种比较反映出的事实在于,哈佛大学的学生一般具有更好的高中学习成绩和更高的SAT分数,做事更有动力,或许还具备其他一些技能和天赋。

- 这种选择偏误可以通过随机分配来消除。遗憾的是,哈佛大学录取办公室并不想用随机数生成器代替录取决策过程。

- 我们必须使用各种类型的大学和学生在常规的申请、录取和入学考试决策中产生的数据来回答选择的选择是否会产生影响。可以用这些数据来模拟我们想在这个背景下进行的随机实验吗?当然,我们不会达到完美状态,但可能会很接近。这项工作的关键在于一个事实,即包括上大学在内的许多决定和选择,都会涉及财务考量、个人环境和时机等因素造成的某些偶然变化。

- 可以从处在转折点附近的申请人组成的样本中挖掘偶然性,这个样本中的申请人可以轻易地走向另外一条道路。被哈佛大学录取的申请人真的会去本地公立大学读书吗?
  我们的朋友, 曾经的麻省理工学院博士生南希 (Nancy)
  就做了这种选择。南希在得克萨斯州长大,得克萨斯州立大学 (University of
  Texas, UT) 是她对应的公立学校。在 Barron
  的排名中,得克萨斯州立大学的旗舰校区奥斯汀分校 (Austin)
  被认为是“极具竞争力”,但它还是比不上哈佛大学。然而,得克萨斯州立大学奥斯汀分校的学费比哈佛大学便宜很多。南希同时被哈佛大学和得克萨斯州立大学奥斯汀分校录取,最终南希选择了得克萨斯州立大学奥斯汀分校,没有选择哈佛大学,因为得克萨斯州立大学奥斯汀分校的招生办急于提高该校区的平均
  SAT
  成绩,为此,它向南希以及其他一些优秀的申请人提供了特别慷慨的财务援助,南希对此欣然接受。

- 南希接受得克萨斯州立大学奥斯汀分校录取并拒绝哈佛大学录取的决策,会对她产生什么影响?虽然南希选择了得克萨斯州立大学奥斯汀分校而不是哈佛大学,她还是做得很出色:如今,她已经是在新英格兰地区另一所常春藤盟校的经济学教授。

- 通过在更大规模的组别之间进行比较,我们有望获取具有一般性的结论。然而,获取大规模样本也不够容易。要进行比较,我们决策中分离偶然性,最重要的是第一步是,使进入私立大学和公立大学的学生之间最明显,最重要的差别保持不变。通过这种方式,我们希望(虽然不能保证)做到其他条件不变。

- 这里用一个小样本数值例子来说明其他条件不变。假设生活中唯一重要的事,至少对你而言,是你的SAT分数和你所在哪里上学。考虑乌玛(Uma)和哈维(Harvey)两个人。这两个人在阅读和数学方面获得的SAT总分都是1400。乌玛去了马萨诸塞州立大学,哈维去了哈佛大学。我们从比较乌玛和哈维的收入开始。因为我们已经假设,除了学校因素之外,影响收入的唯一因素是SAT分数。因此,乌玛和哈维的比较符合其他条件不变假设。

- 当然,现实情况更加复杂。这个简单例子存在一个重大的干扰。乌玛是位年轻女性,哈维是位年轻男性。具有相同教育资质的女性往往比男性赚得少,也许这是因为存在歧视,或者是应为女性要离开劳动力市场生育孩子。哈维比乌玛多赚
  20%,也许是因为哈佛大学优越的教育产生的影响,但是,也许仅仅反映了其他因素造成的男女工资差距。

- 我们想从其他事情中分离出单纯由哈佛大学引起的效应。如果产生影响的其他因素只有性别,那么这种分离就很简单:用
  SAT 总分也是 1400 的哈佛大学学生汉娜 (Hannah)
  代替哈维,然后在乌玛和汉娜之间进行比较。最后,因为我们寻求的是超越个体的一般性结论,所以,要在两个学校中具有同性别、同
  SAT 分数的很多个体之间进行比较。也就是说,我们是在具有相同性别、相同
  SAT
  分数哈佛大学学生和马萨诸塞州立大学学生之间计算平均收入差别。所有这些组别中表现出的哈佛大学学生和马萨诸塞州立大学学生的平均收入差别估计,构成了我们估计哈佛大学教育的因果效应的第一步。这就是控制性别的
  SAT 分数——也就是保持这两个因素不变——的计量经济学的匹配
  (matching)假设在给定性别和 SAT
  分数后,进入哈佛大学和马萨诸塞州立大学的学生平均有着类似的收入潜力,那么这个估计就捕捉到了哈佛大学教育对收入产生的平均因果效应。

### 匹配者

- 再次考察乌玛和哈维的故事:他们都向哈佛大学和马萨诸塞州立大学提交了入学申请,而且也都获得录取。乌玛向哈佛大学提交申请的事实说明,她有动力去那所学校学习,她获得哈佛大学录取的事实则说明,与哈维一样,她有能力在哈佛大学获得成功。至少哈佛大学招生办公室是这么想的,这个办公室可不那么好忽悠。②然而,乌玛选择去更便宜的马萨诸塞州立大学接受教育。她的选择可能源自那些与自身收入潜力关系不大的因素,例如她可能有个很成功的叔叔也在马萨诸塞州立大学学习过,有个最好的朋友去了马萨诸塞州立大学,或是因为乌玛错过了截止日,没能拿到原本可以轻松获得的Rotary
  俱乐部奖学金,从而未能获得进入常春藤盟校的资助。如果这种偶然事件对乌玛和哈维具有决定性影响,那么这两人就是一组很好的匹配。

- 戴尔和克鲁格分析了一个被称为“大学内外”(College and Beyond,
  C&B)的大型数据集。C&B数据集包含数以千计的学生信息,这些学生有的进入了竞争不那么激烈的大学,有的则进入了具有高度选拔性的大学,同时,这个数据集还收集了这些学生在进入大学一年前参加SAT考试时的调查信息,以及1996年大部分学生从大学毕业时收集的信息。分析着重于在1976年进入大学,并在1995年仍在工作的那部分人(大部分成年大学毕业生都在工作),涉及的大学包括宾夕法尼亚大学、普林斯顿大学和耶鲁大学等著名的私立大学,也包括斯沃斯莫尔(Swarth-more)、威廉姆斯(Williams)和欧伯林(Oberlin)等一系列规模较小的私立大学,还有四所公立大学(密歇根大学、北卡罗来纳州立大学、宾夕法尼亚州立大学和俄亥俄州迈阿密大学)。

- 按照申请和录取的学校不同,学生被分为四组。每组中的学生大致会有相同的职业抱负,而且他们也被各自申请的学校的招生官判定为具有类似能力。因此,相比于包含所有学生的那种不受控制的比较,我们应该认为每组内部的比较要更具可比性。

- 虽然组
  A中的学生都做得很好,有着较高的平均收入和较高的私立大学人生学率,但是在组A内部,私立学校产生的收入差异为负:$(110+100)/2-110=-5$,换句话说,差距是$-5000$美元。

  All State 和 Tall
  State,这两所学校都是公立大学,所以他们的收入同样不能揭示出私立大学教育的效应。组C和组D都不含有信息量,因为,我们希望估计的是私立大学产生的处理效应,但是,这两组学生要么都属于处理组,要么都属于控制组。

- 在我们的例子中,组A和组B是起作用的,因为这两组都包含了向公立学校和向私立学校提出入学申请的学生,这些学生也被同一组学校录取。为了获得使用所有可得数据计算出的单一估计值,我们对每组学生对应的估计值进行平均。组A中平均收入之差是-5000美元,组B中平均收入之差是30000美元,二者的平均值是12500美元。这是私立大学教育对平均收入产生影响的一个好的估计值,因为它从很大程度上控制了申请人的选择和能力。

- 对组A和组中得到的处理组——控制组之差简单平均,并不是可从这两组学生身上估计出的唯——个具有良好控制的比较值。例如,我们可以构造一个加权平均值,反映组B包含两名学生、组A包含三名学生的事实。在这个例子中,加权平均值可以计算如下
  $$(\frac{3}{5}×(-5000))+(\frac{2}{5}×30000) = 9000$$
  通过强调规模更大的组别,这种加权平均方式能够更有效地利用数据,而且,也可能因此给出在统计上更精确的私立学校——公立学校毕业生收入差异估计值

- 相比之下,如果使用表中所有九位同学的数据简单比较私立大学和公立大学的学生收入差距,这种幼稚的计算方式会产生一个大得多的差值,达到19500美元。即使将范围限制在组A和组B中的五位同学,不加控制的比较所产生的收入差异也会达到20000美元，不加控制的比较所得差值如此之大,反映出的是选择偏误:向私立大学提出人学申请并被录取的学生,无论去哪里上学都会有更高的收入水平。

- 存在选择偏误的证据来自于在组A和组B之间(而不是分别在组A和组B内部)平均收入的比较。组A中有三分之二的申请面向私立大学,平均收入大约是107000美元。组B中三分之二的申请面向公立大学,平均收入只有45000美元。我们的组内比较揭示出,这种差别在很大程度上与学生选择上哪所大学无关。相反,组间差异由学生的抱负和能力决定,这反映在学生申请大学的决策上,也反映在录取学生的学校上。

## 回归的基本原理

### 条件期望函数（CEF）

- 条件期望函数$E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]$是$\mathrm{X}_i$的函数

- 对于$\mathrm{X}_i$的一个特定值，记作$E\left[\mathrm{Y}_i \mid \mathrm{X}_i=x\right]$，因此

$$
E\left[\mathrm{Y}_i \mid \mathrm{X}_i=x\right]=\int t f_y\left(t \mid \mathrm{X}_i=x\right) d t
$$

或

$$
E\left[\mathrm{Y}_i \mid \mathrm{X}_i=x\right]= \sum_t t P\left(\mathrm{Y}_i=t \mid \mathrm{X}_i=x\right)
$$

- 例如

$$
E[\log (\text { wage }) \mid \text { gender }=\text { man }]=3.05
$$

$$
E[\log (\text { wage }) \mid \text { gender }=\text { woman }]=2.81
$$

$$
E[\log (\text { wage }) \mid \text { gender }=\text { man, race }=\text { white, education }=12]=2.84
$$

**期望迭代法则**

$$
E\left[\mathrm{Y}_i\right]=E\left\{E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]\right\}
$$

证明： $$
\begin{aligned}
E\left\{E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]\right\} & =\int E\left[\mathrm{Y}_i \mid \mathrm{X}_i=u\right] g_x(u) d u \\
& =\int\left[\int t f_y\left(t \mid \mathrm{X}_i=u\right) d t\right] g_x(u) d u \\
& =\iint t f_y\left(t \mid \mathrm{X}_i=u\right) g_x(u) d u d t \\
& =\int t\left[\int f_y\left(t \mid \mathrm{X}_i=u\right) g_x(u) d u\right] d t \\
& =\int t\left[\int f_{x y}(u, t) d u\right] d t \\
& =\int t g_y(t) d t=E\left[\mathrm{Y}_i\right]
\end{aligned}
$$

例如

$$
\begin{aligned}
& E[\log (\text { wage }) \mid \text { gender }=\text { man }] P[\text { gender }=\text { man }] \\
& +E[\log (\text { wage }) \mid \text { gender }=\text { woman }] P[\text { gender }=\text { woman }] \\
& =E[\log (\text { wage })]
\end{aligned}
$$

**CEF分解性质**

$$
\mathbf{Y}_i=E\left[\mathbf{Y}_i \mid \mathrm{X}_i\right]+\varepsilon_i
$$

因此：（1）$E\left[\varepsilon_i \mid \mathrm{X}_i\right]=0$；（2）$\varepsilon_i$与$\mathrm{X}_i$的任何函数都不相关。

证明：

（1） $$
E\left[\varepsilon_i \mid \mathrm{X}_i\right]=E\left[\mathrm{Y}_i-E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right] \mid \mathrm{X}_i\right]=E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]-E\left[\mathrm{Y}_i \mid \mathrm{X}_i \right] = 0
$$

（2） $$
E\left[h\left(\mathrm{X}_i\right) \varepsilon_i\right]=E\left\{h\left(\mathrm{X}_i\right) E\left[\varepsilon_i \mid \mathrm{X}_i\right]\right\}
$$

**CEF预测性质**

令$m(\mathrm{X}_i)$为$\mathrm{X}_i$的任意函数。CEF求解了下式

$$
E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]=\underset{m\left(\mathrm{X}_i\right)}{\arg \min } E\left[\left(\mathrm{Y}_i-m\left(\mathrm{X}_i\right)\right)^2\right]
$$

所以，它是给定$\mathrm{X}_i$时$\mathrm{Y}_i$的最佳预测器。

证明：

$$
\begin{aligned}
\left(\mathrm{Y}_i-m\left(\mathrm{X}_i\right)\right)^2= & \left(\left(\mathrm{Y}_i-E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]\right)+\left(E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]-m\left(\mathrm{X}_i\right)\right)\right)^2 \\
= & \left(\mathrm{Y}_i-E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]\right)^2+2\left(E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]-m\left(\mathrm{X}_i\right)\right) \\
& \times\left(\mathrm{Y}_i-E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]\right)+\left(E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]-m\left(\mathrm{X}_i\right)\right)^2
\end{aligned}
$$

### 线性回归与条件期望函数

对于总体回归模型，最小二乘问题的解

$$
\beta=\underset{b}{\arg \min } E\left[\left(\mathrm{Y}_i-\mathrm{X}_i^{\prime} b\right)^2\right]
$$

一阶条件

$$
E\left[\mathrm{X}_i\left(\mathrm{Y}_i-\mathrm{X}_i^{\prime} b\right)\right]=0
$$

解

$$
\beta=E\left[\mathrm{X}_i \mathrm{X}_i^{\prime}\right]^{-1} E\left[\mathrm{X}_i \mathrm{Y}_i\right]
$$

回归解析公式

$$
\beta_k=\frac{\operatorname{Cov}\left(\mathrm{Y}_i, \tilde{x}_{k i}\right)}{V\left(\tilde{x}_{k i}\right)}
$$

**线性CEF定理**

假设CEF是线性的，那么总体回归函数就是这个CEF

证明：假设$E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]=\mathrm{X}_i^{\prime} \beta^*$，根据CEF分解性质，$E\left[\mathrm{X}_i\left(\mathrm{Y}_i-E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]\right)\right]=0$。把$E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]=\mathrm{X}_i^{\prime} \beta^*$带入，得到$\beta^*=E\left[\mathrm{X}_i \mathrm{X}_i^{\prime}\right]^{-1} E\left[\mathrm{X}_i \mathrm{Y}_i\right]=\beta$。

------------------------------------------------------------------------

**最佳线性预测定理**

函数$\mathrm{X}_i^{\prime} \beta$是给定$\mathrm{X}_i$在最小均方误差（MMSE）意义下$\mathrm{Y}_i$的最佳线性预测函数。

------------------------------------------------------------------------

**回归CEF定理**

函数$\mathrm{X}_i^{\prime} \beta$提供了对$E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]$的最小均方误差（MMSE）线性近似，即

$$
\beta=\underset{b}{\arg \min } E\left\{\left(E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]-\mathrm{X}_i^{\prime} b\right)^2\right\}
$$

证明：

$$
\begin{aligned}
\left(\mathrm{Y}_i-\mathrm{X}_i^{\prime} b\right)^2= & \left\{\left(\mathrm{Y}_i-E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]\right)+\left(E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]-\mathrm{X}_i^{\prime} b\right)\right\}^2 \\
= & \left(\mathrm{Y}_i-E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]\right)^2+\left(E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]-\mathrm{X}_i^{\prime} b\right)^2 \\
& +2\left(\mathrm{Y}_i-E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]\right)\left(E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]-\mathrm{X}_i^{\prime} b\right) 
\end{aligned}
$$

它的一个含义是，可以用$E\left[\mathrm{Y}_i \mid \mathrm{X}_i\right]$代替$\mathbf{Y}_i$本身作为因变量来得到回归系数。

### 矩阵表达

线性方程组

$$
\begin{gathered}
Y_1=X_1^{\prime} \beta+e_1 \\
Y_2=X_2^{\prime} \beta+e_2 \\
\vdots \\
Y_n=X_n^{\prime} \beta+e_n
\end{gathered}
$$

其中

$$
\boldsymbol{Y}=\left(\begin{array}{c}
Y_1 \\
Y_2 \\
\vdots \\
Y_n
\end{array}\right), \quad \boldsymbol{X}=\left(\begin{array}{c}
X_1^{\prime} \\
X_2^{\prime} \\
\vdots \\
X_n^{\prime}
\end{array}\right), \quad \boldsymbol{e}=\left(\begin{array}{c}
e_1 \\
e_2 \\
\vdots \\
e_n
\end{array}\right)
$$

观察到 $\boldsymbol{Y}$ 和 $\boldsymbol{e}$ 是 $n \times 1$ 向量，而
$\boldsymbol{X}$ 是一个 $n \times k$ 矩阵。这 $n$
个方程的系统可以紧凑地写成一个单独的方程：

$$
\boldsymbol{Y}=\boldsymbol{X} \beta+\boldsymbol{e}
$$

样本总和可以用矩阵符号表示。例如：

$$
\begin{aligned}
& \sum_{i=1}^n X_i X_i^{\prime}=\boldsymbol{X}^{\prime} \boldsymbol{X} \\
& \sum_{i=1}^n X_i Y_i=\boldsymbol{X}^{\prime} \boldsymbol{Y}
\end{aligned}
$$

得到

$$
\widehat{\beta}=\left(\boldsymbol{X}^{\prime} \boldsymbol{X}\right)^{-1}\left(\boldsymbol{X}^{\prime} \boldsymbol{Y}\right)
$$

### 无偏性

$$
\begin{aligned}
E\left[\widehat{\beta} \mid X_1, \ldots, X_n\right] & =E\left[\left(\sum_{i=1}^n X_i X_i^{\prime}\right)^{-1}\left(\sum_{i=1}^n X_i Y_i\right) \mid X_1, \ldots, X_n\right] \\
& =\left(\sum_{i=1}^n X_i X_i^{\prime}\right)^{-1} E\left[\left(\sum_{i=1}^n X_i Y_i\right) \mid X_1, \ldots, X_n\right] \\
& =\left(\sum_{i=1}^n X_i X_i^{\prime}\right)^{-1} \sum_{i=1}^n E\left[X_i Y_i \mid X_1, \ldots, X_n\right] \\
& =\left(\sum_{i=1}^n X_i X_i^{\prime}\right)^{-1} \sum_{i=1}^n X_i E\left[Y_i \mid X_i\right] \\
& =\left(\sum_{i=1}^n X_i X_i^{\prime}\right)^{-1} \sum_{i=1}^n X_i X_i^{\prime} \beta \\
& =\beta
\end{aligned}
$$

或者

$$
E[\boldsymbol{Y} \mid \boldsymbol{X}]=\left(\begin{array}{c}
\vdots \\
E\left[Y_i \mid \boldsymbol{X}\right] \\
\vdots
\end{array}\right)=\left(\begin{array}{c}
\vdots \\
X_i^{\prime} \beta \\
\vdots
\end{array}\right)=\boldsymbol{X} \beta
$$

和

$$
E[\boldsymbol{e} \mid \boldsymbol{X}]=\left(\begin{array}{c}
\vdots \\
E\left[e_i \mid \boldsymbol{X}\right] \\
\vdots
\end{array}\right)=\left(\begin{array}{c}
\vdots \\
E\left[e_i \mid X_i\right] \\
\vdots
\end{array}\right)=0
$$

有

$$
\begin{aligned}
\widehat{\beta} & =\left(\boldsymbol{X}^{\prime} \boldsymbol{X}\right)^{-1}\left(\boldsymbol{X}^{\prime}(\boldsymbol{X} \beta+\boldsymbol{e})\right) \\
& =\left(\boldsymbol{X}^{\prime} \boldsymbol{X}\right)^{-1} \boldsymbol{X}^{\prime} \boldsymbol{X} \beta+\left(\boldsymbol{X}^{\prime} \boldsymbol{X}\right)^{-1}\left(\boldsymbol{X}^{\prime} \boldsymbol{e}\right) \\
& =\beta+\left(\boldsymbol{X}^{\prime} \boldsymbol{X}\right)^{-1} \boldsymbol{X}^{\prime} \boldsymbol{e} .
\end{aligned}
$$

因此

$$
\begin{aligned}
E[\widehat{\beta}-\beta \mid \boldsymbol{X}] & =E\left[\left(\boldsymbol{X}^{\prime} \boldsymbol{X}\right)^{-1} \boldsymbol{X}^{\prime} \boldsymbol{e} \mid \boldsymbol{X}\right] \\
& =\left(\boldsymbol{X}^{\prime} \boldsymbol{X}\right)^{-1} \boldsymbol{X}^{\prime} E[\boldsymbol{e} \mid \boldsymbol{X}]=0
\end{aligned}
$$

### OLS估计量的方差

误差方差定义为

$$
\operatorname{var}[\boldsymbol{e} \mid \boldsymbol{X}]=E\left[\boldsymbol{e} \boldsymbol{e}^{\prime} \mid \boldsymbol{X}\right] \stackrel{\text { def }}{=} \boldsymbol{D}
$$

即

$$
\boldsymbol{D}=\operatorname{diag}\left(\sigma_1^2, \ldots, \sigma_n^2\right)=\left(\begin{array}{cccc}
\sigma_1^2 & 0 & \cdots & 0 \\
0 & \sigma_2^2 & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & \sigma_n^2
\end{array}\right)
$$

其中

$$
E\left[e_i^2 \mid \boldsymbol{X}\right]=E\left[e_i^2 \mid X_i\right]=\sigma_i^2
$$

$$
E\left[e_i e_j \mid \boldsymbol{X}\right]=E\left(e_i \mid X_i\right) E\left[e_j \mid X_j\right]=0
$$

对于$n \times r$矩阵$A=A(X)$有

$$
\operatorname{var}\left[\boldsymbol{A}^{\prime} \boldsymbol{Y} \mid \boldsymbol{X}\right]=\operatorname{var}\left[\boldsymbol{A}^{\prime} \boldsymbol{e} \mid \boldsymbol{X}\right]=\boldsymbol{A}^{\prime} \boldsymbol{D} \boldsymbol{A}
$$

因此

$$
\boldsymbol{V}_{\widehat{\beta}}=\operatorname{var}[\widehat{\beta} \mid \boldsymbol{X}]=\boldsymbol{A}^{\prime} \boldsymbol{D} \boldsymbol{A}=\left(\boldsymbol{X}^{\prime} \boldsymbol{X}\right)^{-1} \boldsymbol{X}^{\prime} \boldsymbol{D} \boldsymbol{X}\left(\boldsymbol{X}^{\prime} \boldsymbol{X}\right)^{-1}
$$

其中$\widehat{\beta}=\boldsymbol{A}^{\prime} \boldsymbol{Y}$和$\boldsymbol{A}=\boldsymbol{X}\left(\boldsymbol{X}^{\prime} \boldsymbol{X}\right)^{-1}$

值得一提的是

$$
\boldsymbol{X}^{\prime} \boldsymbol{D} \boldsymbol{X}=\sum_{i=1}^n X_i X_i^{\prime} \sigma_i^2
$$

是$\boldsymbol{X}^{\prime} \boldsymbol{X}$的加权版本

### 饱和模型、主效应和其他回归讨论

**饱和回归模型**是只含有离散解释变量的回归模型，对于解释变量的所有可能值，该模型均纳入一个单独的参数。例如

$$
\mathbf{Y}_i=\alpha+\beta_1 d_{1 i}+\beta_2 d_{2 i}+\cdots+\beta_\tau d_{\tau i}+\varepsilon_i
$$

<br>

如果有两个解释变量——比如一个虚拟变量表示是否大学毕业，另一个虚拟变量表示性别——模型通过包含这两个虚拟变量、它们的乘积和一个常数而饱和。这些虚拟变量的系数被称为**主效应**，而乘积被称为**交互项**。

若用$x_{1 i}$表示大学毕业生，$x_{2 i}$表示女性，那么

$$
\begin{aligned}
& E\left[\mathrm{Y}_i \mid x_{1 i}=0, x_{2 i}=0\right]=\alpha \\
& E\left[\mathrm{Y}_i \mid x_{1 i}=1, x_{2 i}=0\right]=\alpha+\beta_1 \\
& E\left[\mathrm{Y}_i \mid x_{1 i}=0, x_{2 i}=1\right]=\alpha+\gamma \\
& E\left[\mathrm{Y}_i \mid x_{1 i}=1, x_{2 i}=1\right]=\alpha+\beta_1+\gamma+\delta_1
\end{aligned}
$$

由此可得

$$
E\left[\mathrm{Y}_i \mid x_{1 i}, x_{2 i}\right]=\alpha+\beta_1 x_{1 i}+\gamma x_{2 i}+\delta_1\left(x_{1 i} x_{2 i}\right)
$$

## 回归和因果关系

### 条件独立性假设

条件独立性假设（Conditional Independent
Assumption，CIA）是一个核心假设，为回归估计的因果解释提供了理由，这个假设也被称为基于可观测变量的选择，因为保持不变的协变量被假设为已知和可观测的。

现假设教育程度是一个二元决策，用一个虚拟变量$\mathrm{C}_i$表示它。设想两个潜在的收入变量

$$
\text { 潜在结果 }= \begin{cases}Y_{1 i} & \text { 如果 } C_i=1 \\ Y_{0 i} & \text { 如果 } C_i=0\end{cases}
$$

所观测的结果可用潜在结果$\mathrm{Y}_i$的形式表示

$$
Y_i=Y_{0 i}+\left(Y_{1 i}-Y_{0 i}\right) C_i
$$

有

$$
\begin{array}{r}
\underbrace{\mathrm{E}\left[Y_i \mid C_i=1\right]-\mathrm{E}\left[Y_i \mid C_i=0\right]}_{\text {观察到的收人帣距 }}=\underbrace{\mathrm{E}\left[Y_{1 i}-Y_{0 i} \mid C_i=1\right]}_{\text {处理的平均因果效应 }} \\
+\underbrace{\mathrm{E}\left[Y_{0 i} \mid C_i=1\right]-\mathrm{E}\left[Y_{0 i} \mid C_i=0\right]}_{\text {选择性偏误 }}
\end{array}
$$

如果在可观测到的特征$\mathrm{X}_i$为条件下，选择性偏差被消除了，即

$$
\left\{\mathrm{Y}_{0 i}, \mathrm{Y}_{1 i}\right\} \perp \mathrm{C}_i \mid \mathrm{X}_i
$$

给定CIA假设，以$\mathrm{X}_i$为条件，对不同受教育水平的平均收入进行对比可以得出因果关系，即

$$
E\left[\mathrm{Y}_i \mid \mathrm{X}_i, \mathrm{C}_i=1\right]-E\left[\mathrm{Y}_i \mid \mathrm{X}_i, \mathrm{C}_i=0\right]=E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{X}_i\right]
$$

现在拓展到两个以上取值的情况下，假设个人$i$在接受$s$年教育后可能获得的收入为

$$
\mathrm{Y}_{s i} \equiv f_i(s)
$$

它会告诉我们，给定任意值的受教育水平，个体将获得多少收入。换句话说，$f_i(s)$回答了“**如果**某种情况发生，**那么**将会怎么样？”

因此在更一般的框架里，CIA变成了如下表达式，即对于所有的$s$，有

$$
\mathrm{Y}_{\mathrm{s} i} \perp \mathrm{~s}_i \mid \mathrm{X}_i, \text { 对所有的 } s
$$

给定CIA，在$\mathrm{X}_i$的条件下对不同受教育水平的平均收入进行对比，是可以得出因果解释的。即

$$
\begin{aligned}
& E\left[\mathrm{Y}_i \mid \mathrm{X}_i, \mathrm{~s}_i=s\right]-E\left[\mathrm{Y}_i \mid \mathrm{X}_i, \mathrm{~s}_i=s-1\right] \\
& \quad=E\left[f_i(s)-f_i(s-1) \mid \mathrm{X}_i\right]
\end{aligned}
$$

例如，可以通过比较受教育12年和11年的人的收入来认识高中毕业与否的平均因果效应

$$
\begin{aligned}
& E\left[\mathrm{Y}_i \mid \mathrm{X}_i, \mathrm{~s}_i=12\right]-E\left[\mathrm{Y}_i \mid \mathrm{X}_i, \mathrm{~s}_i=11\right] \\
& \quad=E\left[f_i(12) \mid \mathrm{X}_i, \mathrm{~s}_i=12\right]-E\left[f_i(11) \mid \mathrm{X}_i, \mathrm{~s}_i=11\right]
\end{aligned}
$$

因为给定CIA，下式成立

$$
\begin{aligned}
& E\left[f_i(12) \mid \mathrm{X}_i, \mathrm{~s}_i=12\right]-E\left[f_i(11) \mid \mathrm{X}_i, \mathrm{~s}_i=11\right] \\
& \quad=E\left[f_i(12)-f_i(11) \mid \mathrm{X}_i, \mathrm{~s}_i=12\right]
\end{aligned}
$$

根据期望迭代法则，高中毕业的无条件平均因果效应为

$$
\begin{aligned}
& E\left\{E\left[\mathrm{Y}_i \mid \mathrm{X}_i, \mathrm{~s}_i=12\right]-E\left[\mathrm{Y}_i \mid \mathrm{X}_i, \mathrm{~s}_i=11\right]\right\} \\
& \quad=E\left\{E\left[f_i(12)-f_i(11) \mid \mathrm{X}_i\right]\right\} \\
& \quad=E\left[f_i(12)-f_i(11)\right] 
\end{aligned}
$$

同理，高中毕业的平均因果效应为

$$
\begin{aligned}
& \mathrm{E}\left\{\mathrm{E}\left[Y_i \mid X_i, S_i=12\right]-\mathrm{E}\left[Y_i \mid X_i, S_i=11\right] \mid S_i=12\right\} \\
= & \mathrm{E}\left\{\mathrm{E}\left[f_i(12)-f_i(11) \mid X_i\right] \mid S_i=12\right\} \\
= & \mathrm{E}\left[f_i(12)-f_i(11) \mid S_i=12\right]
\end{aligned}
$$

回归提供了一个易于使用的经验策略，自动地把CIA变成因果效应。

路径一是假设$f_i(s)$在$s$上都是线性的，除了可加性误差项外，它对所有人都是相同的。在这种情况下，线性回归是估计$f_i(s)$特征的自然工具。

假设一个线性常数效应因果模型，即

$$
f_i(s)=\alpha+\rho s+\eta_i
$$

将观测值代入，得到

$$
\mathbf{Y}_i=\alpha+\rho \mathrm{s}_i+\eta_i
$$

假设，在给定可观测协变量$\mathrm{X}_i$的向量的情况下，CIA成立，因此

$$
\eta_i=\mathrm{X}_i^{\prime} \gamma+v_i
$$

有

$$
\begin{aligned}
E\left[f_i(s) \mid \mathrm{X}_i, \mathrm{~s}_i\right]=E\left[f_i(s) \mid \mathrm{X}_i\right] & =\alpha+\rho s+E\left[\eta_i \mid \mathrm{X}\right] \\
& =\alpha+\rho s+\mathrm{X}_i^{\prime} \gamma
\end{aligned}
$$

### 遗漏变量偏差公式

遗漏变量偏差 (omitted variables bias, OVB)
公式描述了拥有不同控制变量集的模型回归估计值之间的关系。这个重要的公式通常是由这样一个观念驱动的：较长的回归具有因果解释，而较短的回归则没有因果解释。因此，较短回归中所含变量的系数被认为是有偏的。事实上，OVB
公式是适用于短回归和长回归的系数向量之间的一种联系机制，而不论长回归是否具有因果解释。

假设学校教育回归中的相关控制变量集可以被归结为家庭背景、智力和个人志向的组合。让我们用向量$A_i$来表示这些特定因素，将其简称为
“能力”。控制能力因素之后，工资对受教育年限$\mathrm{S}_i$的回归可以写成:

$$
\mathrm{Y}_i=\alpha+\rho \mathrm{s}_i+A_i^{\prime} \gamma+e_i
$$

**OVB公式**

$$
\frac{\operatorname{Cov}\left(\mathrm{Y}_i, \mathrm{~s}_i\right)}{V\left(\mathrm{~s}_i\right)}=\rho+\gamma^{\prime} \delta_{A s}
$$

其中$\delta_{A s}$是$A_i$元素对$\mathrm{S}_i$回归所得的系数向量。即

- **短回归等于长回归加上遗漏变量效应乘以遗漏变量对所包含变量的回归。**

<br>

案例：使用全国青年纵向调查(National Longitudinal Survey of Youth,
NLSY)的数据，表中的前三个条目显示,当家庭背景变量(本例中是父母的受教育程度)以及一些基本人口统计特征(年龄、种族、人口普查居住地)作为控制变量被纳入时,学校教育系数从0.132下降到0.114。将军队资格测试(AFQT)分数作为代理变量来进一步控制个人能力,可将学校教育系数降低到0.087(AFQT是军队用来选拔士兵的测试)。OVB公式告诉我们,这些减少是由于增加的控制变量与工资和学校教育都呈正相关。

OVB公式的重要性源于这样一个事实:如果你认为没有遗漏变量偏差,那么通常你也会说得到的回归就是你想要的回归。而你想要的回归通常会有一个因果解释。

值得考虑的是,CIA什么时候最有可能为实证工作提供可信的基础。最好的情况是在某种实验(可能是自然实验)中,以$\mathrm{X}_i$为条件随机分配$\mathrm{S}_i$。

案例：Black等(2003)对失业工人强制性再培训计划的研究即为一例。令这项研究的作者感兴趣的是,再培训计划能否在之后成功地提高收入。他们利用了这样一个事实:他们所研究的项目中的培训资格是根据个人特征以及过去的失业情况和工作经历来确定的。工人们根据这些特点被分成不同的群体。虽然一些工人群体中的某些人没有资格接受培训,但其他群体中的工人如果没有工作就必须接受培训。当这些强制性培训组的工人人数超过培训名额时,培训机会则是通过抽签分配的。因此,以将工人分配到各组的协变量为条件,培训要求此时是随机分配的。对是否受到培训的虚拟变量,以及个人特征、过去的失业情况变量和用于对工人进行分类的工作经历变量进行回归,似乎很有可能提供对培训因果效应的可靠估计值。

### 不良的控制

不良的控制变量本身会是我们所设想的实验的结果变量,也就是说,不良的控制变量也可能是因变量。良好的控制变量,是我们可以考虑的当感兴趣的回归元选定时(它们)也已经固定的变量。

假设我们对大学学历影响收入的问题感兴趣,同时设想我们可以从事两种职业(白领和蓝领)之一。大学学历显然是高薪白领工作的“入场券”。因此,职业是否应该被视为工资对受教育年限回归中被遗漏的变量呢?毕竟,职业与教育和收入都高度相关。也许,最好还是在一类职业中观察大学(学历)对工资的影响,比如仅看白领。该问题的争议在于,一旦我们承认大学学历影响职业这一事实,即使大学学历是随机分配的,同类职业中大学学历身份(与否)的工资比较,也不再像苹果与苹果的比较那样具有可比性了。

设$\mathrm{W}_i$为虚拟变量,表示是否为白领,$\mathbf{Y}_i$表示收入。这些变量的实现取决于大学学历身份(用$\mathrm{C}_i$表示)以及与$\mathrm{C}_i$挂钩的潜在结果。

$$
\mathrm{Y}_i=\mathrm{C}_i \mathrm{Y}_{1 i}+\left(1-\mathrm{C}_i\right) \mathrm{Y}_{0 i}
$$

$$
\mathrm{W}_i=\mathrm{C}_i \mathrm{~W}_{1 i}+\left(1-\mathrm{C}_i\right) \mathrm{W}_{0 i}
$$

其中, $\mathrm{C}_i=1$表示拥有大学学历,
$\mathrm{C}_i=0$表示未取得大学学历,
$\left\{\mathrm{Y}_{1 i}, \mathrm{Y}_{0 i}\right\}$表示潜在的收入,
$\left\{\mathbf{w}_{1 i}, w_{0 i}\right\}$表示潜在的白领身份与否。我们假设$\mathrm{C}_i$是随机分配的,
因此它独立于所有可能的结果。

$$
E\left[\mathrm{Y}_i \mid \mathrm{C}_i=1\right]-E\left[\mathrm{Y}_i \mid \mathrm{C}_i=0\right]=E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i}\right]
$$

$$
E\left[\mathbf{W}_i \mid \mathrm{C}_i=1\right]-E\left[\mathbf{W}_i \mid \mathrm{C}_i=0\right]=E\left[\mathbf{W}_{1 i}-\mathrm{W}_{0 i}\right]
$$

不良的控制意味着以$\mathrm{W}_i$为条件的收入比较不具有因果解释。以白领身份为条件，考虑大学毕业生和非大学毕业生的平均收入差异。我们可以在一个包含$\mathrm{W}_i$的回归模型中计算这个值，或者在$\mathrm{w}_i=1$的样本中通过$\mathrm{Y}_i$对$\mathrm{C}_i$进行回归来计算。后一种情况下的估计值是在$\mathrm{w}_i=1$的条件下$\mathrm{C}_i=1$和$\mathrm{C}_i=0$的平均值之差。

$$
\begin{aligned}
& E\left[\mathrm{Y}_i \mid \mathrm{W}_i=1, \mathrm{C}_i=1\right]-E\left[\mathrm{Y}_i \mid \mathrm{W}_i=1, \mathrm{C}_i=0\right] \\
& \quad=E\left[\mathrm{Y}_{1 i} \mid \mathrm{W}_{1 i}=1, \mathrm{C}_i=1\right]-E\left[\mathrm{Y}_{0 i} \mid \mathrm{W}_{0 i}=1, \mathrm{C}_i=0\right]
\end{aligned}
$$

由$\left\{\mathrm{Y}_{1 i}, \mathrm{~W}_{1 i}, \mathrm{Y}_{0 i}, \mathrm{~W}_{0 i}\right\}$和$\mathrm{C}_i$的联合独立性可得

$$
\begin{aligned}
& E\left[\mathrm{Y}_{1 i} \mid \mathrm{W}_{1 i}=1, \mathrm{C}_i=1\right]-E\left[\mathrm{Y}_{0 i} \mid \mathrm{W}_{0 i}=1, \mathrm{C}_i=0\right] \\
& \quad=E\left[\mathrm{r}_{1 i} \mid \mathrm{W}_{1 i}=1\right]-E\left[\mathrm{Y}_{0 i} \mid \mathrm{W}_{0 i}=1\right]
\end{aligned}
$$

这个表达式点出了不良控制问题的本质

$$
\begin{aligned}
& E\left[\mathrm{Y}_{1 i} \mid \mathrm{W}_{1 i}=1\right]-E\left[\mathrm{Y}_{0 i} \mid \mathrm{W}_{0 i}=1\right] \\
& \quad=\underbrace{E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{W}_{1 i}=1\right]}_{\text {因果效应}}+\underbrace{\left\{E\left[\mathrm{Y}_{0 i} \mid \mathrm{W}_{1 i}=1\right]-E\left[\mathrm{Y}_{0 i} \mid \mathrm{W}_{0 i}=1\right]\right\}}_{\text {选择性偏误}}
\end{aligned}
$$

换句话说，以是否从事白领工作为条件，有无大学学历的工资差异，等于那些$\mathrm{w}_{1 i}=1$的人
(拥有大学学历且为白领工人)
的大学学历的因果效应加上选择性偏差这一项，该偏差项反映了这样一个事实，大学学历改变了白领工人群组的构成。

不良控制问题的第二种情况,涉及代理控制变量(proxy
control),也就是说,回归包含了一些或许可以部分控制遗漏变量的变量,但它们自身也会受到我们所感兴趣的回归元的影响。

$$
\mathrm{Y}_i=\alpha+\rho \mathrm{s}_i+\gamma a_i+e_i
$$

我们可以把$a_i$看作在作出任何相关的教育选择之前(假设每个人都完成了八年级学业)一个衡量八年级学生先天能力的智商分数。但遗憾的是它的数据是不可得的。不过,我们还可以找到第二种能力指标(比如,用于筛选求职者的测试分数),这个指标是在人们完成学业之后收集的。我们不妨把它称为后天能力变量$a_{l i}$。一般来说,相对于先天能力,学校教育会提高后天能力。

$$
a_{l i}=\pi_0+\pi_1 \mathrm{~s}_i+\pi_2 a_i
$$

我们可称学校教育和先天能力都提高了后天或(其他)可观测的能力。加入代理变量的回归为

$$
\mathrm{Y}_i=\left(\alpha-\gamma \frac{\pi_0}{\pi_2}\right)+\left(\rho-\gamma \frac{\pi_1}{\pi_2}\right) \mathrm{s}_i+\frac{\gamma}{\pi_2} a_{l i}+e_i
$$

使用随感兴趣的变量而增加的代理控制变量,会产生低于预期效应的系数。

## 异质性和非线性

### 回归与匹配

匹配策略的一个吸引人的特点是,它们通常伴随着一个明确的条件独立假设的表述,并需要给出匹配估计值的因果解释。同时,我们已经看到,回归系数的因果解释是基于完全相同的假设的。换句话说,匹配和回归都是控制策略。由于两种策略的因果推理的核心假设相同,所以匹配是否或在多大程度上真正不同于回归是值得一问的。我们的观点是,回归可以作为一种特殊的加权匹配估计量,因此回归和匹配估计值之间的差异不太可能具有较大的实证重要性。

匹配估计量通常是由协变量定义的分组间对比或比较的加权平均值。比如在参军的例子中,我们用虚拟变量$\mathrm{D}_i$来表示退伍军人身份这个离散回归元。由于处理变量只取两个值,所以我们可以使用$\mathbf{Y}_{1 i}$和$\mathbf{Y}_{0 i}$表示潜在结果。有

$$
\begin{aligned}
& E\left[\mathrm{Y}_i \mid \mathrm{D}_i=1\right]-E\left[\mathrm{Y}_i \mid \mathrm{D}_i=0\right] \\
& \quad=E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_i=1\right]+\left\{E\left[\mathrm{Y}_{0 i} \mid \mathrm{D}_i=1\right]-E\left[\mathrm{Y}_{0 i} \mid \mathrm{D}_i=0\right]\right\}
\end{aligned}
$$

假设CIA条件为$\left\{\mathrm{Y}_{0 i}, \mathrm{Y}_{1 i}\right\} \perp \mathrm{D}_i \mid \mathrm{X}_i$，因此处理组的平均处理效应为

$$
\begin{aligned}
\delta_{\text {TOT }} & \equiv \mathrm{E}\left[Y_{1 i}-Y_{0 i} \mid D_i=1\right] \\
& =\mathrm{E}\left\{\mathrm{E}\left[Y_{1 i}-Y_{0 i} \mid X_i, D_i=1\right] \mid D_i=1\right\} \\
& =\mathrm{E}\left\{\mathrm{E}\left[Y_{1 i} \mid X_i, D_i=1\right]-\mathrm{E}\left[Y_{0 i} \mid X_i, D_i=1\right] \mid D_i=1\right\}
\end{aligned}
$$

根据CIA条件有

$$
E\left[\mathrm{Y}_{0 i} \mid \mathrm{X}_i, \mathrm{D}_i=0\right]=E\left[\mathrm{Y}_{0 i} \mid \mathrm{X}_i, \mathrm{D}_i=1\right]
$$

因此

$$
\begin{aligned}
\delta_{\text {TOT }} & =E\left\{E\left[\mathrm{Y}_{1 i} \mid \mathrm{X}_i, \mathrm{D}_i=1\right]-E\left[\mathrm{Y}_{0 i} \mid \mathrm{X}_i, \mathrm{D}_i=0\right] \mid \mathrm{D}_i=1\right\} \\
& =E\left[\delta_X \mid \mathrm{D}_i=1\right]
\end{aligned}
$$

其中

$$
\delta_X \equiv E\left[\mathrm{Y}_i \mid \mathrm{X}_i, \mathrm{D}_i=1\right]-E\left[\mathrm{Y}_i \mid \mathrm{X}_i, \mathrm{D}_i=0\right]
$$

它是在$\mathrm{X}_i$的每一个值上根据退伍军人身份与否计算的平均收入之差。

在离散情况下,匹配估计量可以写成

$$
E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_i=1\right]=\sum_x \delta_x P\left(\mathrm{X}_i=x \mid \mathrm{D}_i=1\right)
$$

其中$P\left(\mathrm{X}_i=x \mid \mathrm{D}_i=1\right)$是$\mathrm{D}_i=1$时$\mathrm{X}_i$的概率质量函数。

我们可以很容易地构建无条件平均处理效应

$$
\begin{aligned}
\delta_{A T E} & =E\left\{E\left[\mathrm{Y}_{1 i} \mid \mathrm{X}_i, \mathrm{D}_i=1\right]-E\left[\mathrm{Y}_{0 i} \mid \mathrm{X}_i, \mathrm{D}_i=0\right]\right\} \\
& =\sum_x \delta_x P\left(\mathrm{X}_i=x\right)=E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i}\right]
\end{aligned}
$$

回归估计方程为

$$
\mathbf{Y}_i=\sum_x d_{i x} \alpha_x+\delta_R \mathrm{D}_i+e_i
$$

回归估计值和匹配估计值相似的原因是,回归也可以被视为一种匹配估计量:回归被估量与匹配被估量的区别,仅在于用来将协变量特定的效应$\delta_X$组合为单个平均效应的权重。匹配用协变量的分布作为权重得到加权的匹配估计量，回归用估计系数的方差作为权重得到新的回归估计量。

$$
\begin{aligned}
\delta_R & =\frac{\operatorname{Cov}\left(\mathrm{Y}_i, \tilde{\mathrm{D}}_i\right)}{V\left(\tilde{\mathrm{D}}_i\right)} \\
& =\frac{E\left[\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right) \mathrm{Y}_i\right]}{E\left[\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right)^2\right]} \\
& =\frac{E\left\{\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right) E\left[\mathrm{Y}_i \mid \mathrm{D}_i, \mathrm{X}_i\right]\right\}}{E\left[\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right)^2\right]}
\end{aligned}
$$

进一步

$$
E\left[\mathrm{Y}_i \mid \mathrm{D}_i, \mathrm{X}_i\right]=E\left[\mathrm{Y}_i \mid \mathrm{D}_i=0, \mathrm{X}_i\right]+\delta_X \mathrm{D}_i
$$

代入得到

$$
\begin{aligned}
& E\left\{\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right) E\left[\mathrm{Y}_i \mid \mathrm{D}_i, \mathrm{X}_i\right]\right\} \\
& \quad=E\left\{\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right) E\left[\mathrm{Y}_i \mid \mathrm{D}_i=0, \mathrm{X}_i\right]\right\} \\
& \quad \quad+E\left\{\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right) \mathrm{D}_i \delta_X\right\} 
\end{aligned}
$$

右端的第一项是零,第二项化简为:

$$
E\left\{\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right) \mathrm{D}_i \delta_X\right\}=E\left\{\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right)^2 \delta_X\right\}
$$

因此

$$
\begin{aligned}
\delta_R & =\frac{E\left[\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right)^2 \delta_X\right]}{E\left[\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right)^2\right]} \\
& =\frac{E\left\{E\left[\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right)^2 \mid \mathrm{X}_i\right] \delta_X\right\}}{E\left\{E\left[\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right)^2 \mid \mathrm{X}_i\right]\right\}}=\frac{E\left[\sigma_D^2\left(\mathrm{X}_i\right) \delta_X\right]}{E\left[\sigma_D^2\left(\mathrm{X}_i\right)\right]}
\end{aligned}
$$

其中

$$
\sigma_D^2\left(\mathrm{X}_i\right) \equiv E\left[\left(\mathrm{D}_i-E\left[\mathrm{D}_i \mid \mathrm{X}_i\right]\right)^2 \mid \mathrm{X}_i\right]
$$

它是给定$\mathrm{X}_i$时的$\mathrm{D}_i$条件方差。由于

$$
\sigma_D^2\left(X_i\right)=\mathrm{P}\left(D_i=1 \mid X_i\right)\left(1-\mathrm{P}\left(D_i=1 \mid X_i\right)\right)
$$

所以

$$
\delta_R=\frac{\sum_x \delta_x\left[P\left(\mathrm{D}_i=1 \mid \mathrm{X}_i=x\right)\left(1-P\left(\mathrm{D}_i=1 \mid \mathrm{X}_i=x\right)\right)\right] P\left(\mathrm{X}_i=x\right)}{\sum_x\left[P\left(\mathrm{D}_i=1 \mid \mathrm{X}_i=x\right)\left(1-P\left(\mathrm{D}_i=1 \mid \mathrm{X}_i=x\right)\right)\right] P\left(\mathrm{X}_i=x\right)}
$$

相比之下,处理组的处理效应的匹配估计量可以写为:

$$
\begin{aligned}
E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_i=1\right] & =\sum_x \delta_x P\left(\mathrm{X}_i=x \mid \mathrm{D}_i=1\right) \\
& =\frac{\sum_x \delta_x P\left(\mathrm{D}_i=1 \mid \mathrm{X}_i=x\right) P\left(\mathrm{X}_i=x\right)}{\sum_x P\left(\mathrm{D}_i=1 \mid \mathrm{X}_i=x\right) P\left(\mathrm{X}_i=x\right)}
\end{aligned}
$$

利用下面这一事实

$$
P\left(\mathrm{X}_i=x \mid \mathrm{D}_i=1\right)=\frac{P\left(\mathrm{D}_i=1 \mid \mathrm{X}_i=x\right) \cdot P\left(\mathrm{X}_i=x\right)}{P\left(\mathrm{D}_i=1\right)}
$$

### 使用倾向评分控制协变量

倾向评分定理指出,如果潜在结果独立于以多元协变量向量$\mathrm{X}_i$为条件的处理状态,那么潜在结果就独立于以协变量标量函数(即倾向评分)为条件的处理状态,倾向评分的定义为$p\left(X_i\right) \equiv \mathrm{E}\left[D_i \mid X_i\right]=\mathrm{P}\left[D_i=1 \mid X_i\right]$。正式地,我们有以下定理:

**定理：倾向评分定理**

假设CIA
成立,从而使$\left\{\mathrm{Y}_{0 i}, \mathrm{Y}_{1 i}\right\} \perp \mathrm{D}_i \mid \mathrm{X}_i$。那么有

$$
\left\{\mathrm{Y}_{0 i}, \mathrm{Y}_{1 i}\right\} \perp \mathrm{D}_i \mid p\left(\mathrm{X}_i\right)
$$

证明：相当于证明对于$j=0,1$，$P\left[\mathrm{D}_i=1 \mid \mathrm{Y}_{j i}, p\left(\mathrm{X}_i\right)\right]$不依赖于$\mathbf{Y}_{j i}$即可

$$
\begin{aligned}
P\left[\mathrm{D}_i=1 \mid \mathrm{Y}_{j i}, p\left(\mathrm{X}_i\right)\right] & =E\left[\mathrm{D}_i \mid \mathrm{Y}_{j i}, p\left(\mathrm{X}_i\right)\right] \\
& =E\left\{E\left[\mathrm{D}_i \mid \mathrm{Y}_{j i}, p\left(\mathrm{X}_i\right), \mathrm{X}_i\right] \mid \mathrm{Y}_{j i}, p\left(\mathrm{X}_i\right)\right\} \\
& =E\left\{E\left[\mathrm{D}_i \mid \mathrm{Y}_{j i}, \mathrm{X}_i\right] \mid \mathrm{Y}_{j i}, p\left(\mathrm{X}_i\right)\right\} \\
& =E\left\{E\left[\mathrm{D}_i \mid \mathrm{X}_i\right] \mid \mathrm{Y}_{j i}, p\left(\mathrm{X}_i\right)\right\}, \text {基于CIA条件}
\end{aligned}
$$

即

$$
E\left\{E\left[\mathrm{D}_i \mid \mathrm{X}_i\right] \mid \mathrm{Y}_{j i}, p\left(\mathrm{X}_i\right)\right\}=E\left\{p\left(\mathrm{X}_i\right) \mid \mathrm{Y}_{j i}, p\left(\mathrm{X}_i\right)\right\}
$$

它其实就是$p\left(\mathrm{X}_i\right)$。

在实践中,倾向评分定理通常经两个步骤进行估计 - 首先,使用诸如 Logit 或
Probit 的某种参数模型估计$p\left(\mathrm{X}_i\right)$ -
然后,通过匹配第一步的估计评分，计算处理效应的估计值

直觉上来说，如果两个个体A和B有相同的倾向得分（例如$0.6$），但一个在处理组一个在控制组，如果满足条件独立性假设，那么其观测结果的差异就可以归因于处理。

给定一个估计$p\left(\mathrm{X}_i\right)$的方案，可以构建平均处理效应的估计量

$$
\begin{aligned}
E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i}\right] & =E\left[\frac{\mathrm{Y}_i \mathrm{D}_i}{p\left(\mathrm{X}_i\right)}-\frac{\mathrm{Y}_i\left(1-\mathrm{D}_i\right)}{1-p\left(\mathrm{X}_i\right)}\right] \\
& =E\left[\frac{\left(\mathrm{D}_i-p\left(\mathrm{X}_i\right)\right) \mathrm{Y}_i}{p\left(\mathrm{X}_i\right)\left(1-p\left(\mathrm{X}_i\right)\right)}\right]
\end{aligned}
$$

类似的，处理组的处理效应为

$$
E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_i=1\right]=E\left[\frac{\left(\mathrm{D}_i-p\left(\mathrm{X}_i\right)\right) \mathrm{Y}_i}{\left(1-p\left(\mathrm{X}_i\right)\right) P\left(\mathrm{D}_i=1\right)}\right]
$$

回归估计量为

$$
\delta_R=\frac{E\left[\left(\mathrm{D}_i-p\left(\mathrm{X}_i\right)\right) \mathrm{Y}_i\right]}{E\left[p\left(\mathrm{X}_i\right)\left(1-p\left(\mathrm{X}_i\right)\right)\right]}
$$

### 倾向评分法与回归

这里使用 Dehejia 和
Wahba(1999)中相同的国家支持工作示范项目(NSW)样本来说明回归和倾向评分匹配之间的对比。NSW是一个20世纪70年代中期的项目,为身为弱势短期零工的接受者提供工作经验。NSW是在一项随机试验中做出的评估,这在当时颇不寻常。Lalonde
(1986)的开创性分析将 NSW
随机试验研究的结果与来自美国收入动态面板数据(PSID)和CPS的非实验对照组的计量经济学结果进行了比较。因为看似合理的非实验方法产生了泛泛的结果,而且其中许多结果与实验估计值相去甚远,他颇为悲观地放弃了这个研究。此外,Lalonde认为,一名对于随机试验结果所知不多的客观的研究者,不太可能选到最佳的计量经济学设定和观测对照组。

Dehejia 和
Wahba(1999)对Lalonde(1986)的发现进行了引人注目的二次研究,他们发现,通过将
NSW
处理组与使用倾向评分选择的观测对照组相匹配,得出的结论是接近NSW的实验结果的。他们用不同的对照组证明了这一点。继Dehejia和Wahba(1999)之后,我们再次考察了两个CPS对照组,第一个是基本上未经选择的样本(CPS-1),第二个是从最近失业的人群中选择的一个更窄的对照组(CPS-3)。
