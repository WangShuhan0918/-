## 管用的工具变量

## 工具变量和因果关系

我们给出一个不变效应架构，作为学校教育和工资之间因果关系的框架，我们可以写出潜在结果：

$$
\mathrm{Y}_{s i} \equiv f_i(s)
$$

以及

$$
f_i(s)=\alpha+\rho s+\eta_i
$$

我们假设存在一个控制变量向量$A_i$，即所谓的“能力”，它能够提供一个基于可观测变量选择的表达式：

$$
\eta_i=A_i^{\prime} \gamma+v_i
$$

目前假设，变量$A_i$是$\eta_i$和$\mathrm{s}_i$相关的唯一原因，所以：

$$
E\left[\mathrm{~s}_i v_i\right]=0
$$

如果$A_i$可以被观测到，那么我们将乐于把它纳入工资对学校教育的回归中，使用长回归

$$
\mathrm{Y}_i=\alpha+\rho \mathrm{s}_i+A_i^{\prime} \gamma+v_i
$$

当研究者可以得到一个变量(即“工具”，我们将其记为$\mathrm{Z}_i$)，该变量与感兴趣的因果变量$s_i$相关，但与因变量的任何其他决定因素不相关。这里，所谓的“与因变量的任何其他决定因素不相关”，就是在说$\operatorname{Cov}\left(\eta_i, Z_i\right)=0$,
或者也可以这样说，$\mathrm{Z}_i$与$A_i$和$v_i$都不相关。这种说法被称为排他性约束(exclusion
restriction)，因为$\mathrm{Z}_i$可以说成是被排除在感兴趣的因果模型之外的。

给定排他性约束，有

$$
\rho=\frac{\operatorname{Cov}\left(\mathrm{Y}_i, \mathrm{Z}_i\right)}{\operatorname{Cov}\left(\mathrm{s}_i, \mathrm{Z}_i\right)}=\frac{\operatorname{Cov}\left(\mathrm{Y}_i, \mathrm{Z}_i\right) / V\left(\mathrm{Z}_i\right)}{\operatorname{Cov}\left(\mathrm{s}_i, \mathrm{Z}_i\right) / V\left(\mathrm{Z}_i\right)}
$$

工具变量被估量基于第一阶段不为零的概念，而这是可以通过检查数据得出来的。一般来说，如果第一阶段与零值仅略有显著差异，那么所得到的工具变量估计就不太可能提供什么信息。

上式要得到因果效应，需满足下述条件

- 首先，工具变量必须对 sᵢ 有明显的影响，这是第一阶段。

- 其次，$\mathbf{Y}_i$和$\mathrm{Z}_i$之间存在关系的唯一原因源自第一阶段。目前，我们称这第二个假设为排他性约束，不过我们将在讨论具有异质效应的模型时看到，这个假设实际上有两个部分：第一个部分是，工具变量如同随机分配一般好；第二个部分是，除非通过第一阶段的渠道，工具变量对结果没有影响。

Angrist和Krueger（1991）所描述的故事的数学表达式来自第一阶段和简化形式的回归方程，具体如下：

$$
\begin{aligned}
\mathrm{s}_i & =\mathrm{X}_i^{\prime} \pi_{10}+\pi_{11} \mathrm{Z}_i+\xi_{1 i} \\
\mathrm{Y}_i & =\mathrm{X}_i^{\prime} \pi_{20}+\pi_{21} \mathrm{Z}_i+\xi_{2 i}
\end{aligned}
$$

在 Angrist 和 Krueger (1991)
中，工具变量$\mathrm{Z}_i$是出生季度（或表示出生季度的虚拟变量），协变量是出生年份和出生地情况的虚拟变量。用
SEM 的语言来说，这两个方程中的因变量被称为内生变量 (endogenous
variables, 在联立方程组内部共同确定)，而右端的变量被称为外生变量
(exogenous variables 在联立方程组外部确定)。

因此

$$
\rho=\frac{\pi_{21}}{\pi_{11}}=\frac{\operatorname{Cov}\left(\mathrm{Y}_i, \tilde{z}_i\right)}{\operatorname{Cov}\left(\mathrm{s}_i, \tilde{z}_i\right)}
$$

#### 两阶段最小二乘

把第一阶段方程代入，得到

$$
\begin{aligned}
\mathrm{Y}_i & =\alpha^{\prime} \mathrm{X}_i+\rho\left[\mathrm{X}_i^{\prime} \pi_{10}+\pi_{11} \mathrm{Z}_i+\xi_{1 i}\right]+\eta_i \\
& =\mathrm{X}_i^{\prime}\left[\alpha+\rho \pi_{10}\right]+\rho \pi_{11} \mathrm{Z}_i+\left[\rho \xi_{1 i}+\eta_i\right] \\
& =\mathrm{X}_i^{\prime} \pi_{20}+\pi_{21} \mathrm{Z}_i+\xi_{2 i}
\end{aligned}
$$

稍加整理得到

$$
\mathbf{Y}_i=\alpha^{\prime} \mathbf{X}_i+\rho\left[\mathrm{X}_i^{\prime} \pi_{10}+\pi_{11} \mathrm{Z}_i\right]+\xi_{2 i}
$$

给定一个随机样本,第一阶段拟合值的一致性估计由下式给出

$$
\hat{s}_i=\mathrm{X}_i^{\prime} \hat{\pi}_{10}+\hat{\pi}_{11} \mathrm{Z}_i
$$

因此$\hat{S}_i$的系数$\rho$被称为2SLS估计量。2SLS估计值也可以通过以下“第二阶段方程”的OLS估计来构建

$$
\mathrm{Y}_i=\alpha^{\prime} \mathrm{X}_i+\rho \hat{\mathrm{s}}_i+\left[\eta_i+\rho\left(\mathrm{s}_i-\hat{s}_i\right)\right]
$$

上式之所以被称为2SLS,是因为它可以分两步完成。由于协变量及第一阶段拟合值与$\eta_i$和$\left(\mathrm{s}_i-\hat{s}_i\right)$都不相关,由此得到的$\rho$的估计值是一致性估计。

2SLS虽名为“两步”,但我们通常不会用两步来构建2SLS估计值。首因是产生的标准误是错误的。直观来看,在协变量的条件下,2SLS只保留了由准实验变化产生的$S_i$的变化,即由工具变量$\mathrm{Z}_i$产生的$S_i$的变化。

在多工具变量模型中,2SLS通过将多个工具变量合并成一个工具变量来实现这一点。例如,假设我们有三个工具变量$\mathbf{Z}_{1 i}$、$\mathbf{Z}_{2 i}$和$\mathbf{Z}_{3 i}$。在Angrist
和 Krueger
(1991)的应用中,这些分别为是否在第一季度、第二季度和第三季度出生的虚拟变量。那么,第一阶段方程就变成:

$$
\mathrm{s}_i=\mathrm{X}_i^{\prime} \pi_{10}+\pi_{11} \mathrm{Z}_{1 i}+\pi_{12} \mathrm{Z}_{2 i}+\pi_{13} \mathrm{Z}_{3 i}+\xi_{1 i}
$$

#### 瓦尔德估计量（Wald Estimator）

最简单的工具变量估计量应该是使用单一虚拟工具变量来估计只有一个内生回归元且没有协变量的模型。由于没有协变量,所以这个因果回归模型为:

$$
\mathrm{Y}_i=\alpha+\rho \mathrm{s}_i+\eta_i
$$

令$\mathrm{Z}_i$是一个虚拟变量,以概率$p$等于1，可以得到

$$
\operatorname{Cov}\left(\mathrm{Y}_i, \mathrm{Z}_i\right)=\left\{E\left[\mathrm{Y}_i \mid \mathrm{Z}_i=1\right]-E\left[\mathrm{Y}_i \mid \mathrm{Z}_i=0\right]\right\} p(1-p)
$$

由此可以得到瓦尔德估计量（Wald Estimator）

$$
\rho=\frac{E\left[\mathrm{Y}_i \mid \mathrm{Z}_i=1\right]-E\left[\mathrm{Y}_i \mid \mathrm{Z}_i=0\right]}{E\left[\mathrm{~s}_i \mid \mathrm{Z}_i=1\right]-E\left[\mathrm{~s}_i \mid \mathrm{Z}_i=0\right]}
$$

或者直接利用

$$
E\left[\mathrm{Y}_i \mid \mathrm{Z}_i\right]=\alpha+\rho E\left[\mathrm{~s}_i \mid \mathrm{Z}_i\right]
$$

可以得到上述结果。

瓦尔德公式为消除OVB的工具变量策略提供了一个引人注目的且明确的实施路径。引发工具变量因果效应估计的主要理由是,因变量和工到变量之间存在任何关系的唯一原因是工具变量对感兴趣的因果变量的效应。因此,
在虚拟工具变量的情形中,用相应的第一阶段的均值差来除——或重新缩放(rescale)——简化形式的均值差,似乎是很自然的。

Angrist(1990)关于越南战争时期服役对退伍军人收入影响的研究也显示了瓦尔德估计量的作用。在20世纪60年代和70年代初,年轻的美国男性面临着被征召入伍的可能。对美国征兵政策公平性的担忧导致了1970年的征兵抽签制度,该制度被用来确定征兵的先后次序。因此,**关于军人身份的一个大有可为的工具变量就是应征资格,因为这是由生日来抽签决定的。具体来说,1970—1972年,给19岁的青年群组中的每个出生日期随机分配随机序列号(RSN)。抽签号码低于临界值的男性有资格应征入伍,而抽签号码高于临界值的男性则不能应征入伍。在实践中,许多符合征兵条件的男子仍然因为健康或其他原因而被免除兵役,而许多被免除兵役的男子却自愿参军。因此,军人身份并不完全由随机的应征资格决定,但应征资格提供了一个与越南战争时期军人身份高度相关的虚拟工具变量。**

在那些有可能于1970年抽签中被选中的白人男性里,其应征资格显然与抽签后几年的较低收入有关。表的第2列记录了这一点,它报告了随机应征资格状态对社会保障应税收入的影响。第(1)列显示的是平均年收入,以便比较。对于1950年出生的男性来说,应征资格对1971年的收入有显著的负面影响,当时这些男性大多刚刚开始服兵役,也许更令人惊讶的负面影响来自十年后(1981年)的比较。相比之下,**没有证据表明在1969年应征资格和收入之间存在关联。1969年是对1950年出生的男性进行抽签,但出生于1950年的人实际上还没有应征入伍。**

因为**应征资格是随机分配的,所以第(2)列的估计值代表了应征资格对收入的因果效应的说法似乎是没有争议的。**从应征资格效应到军人身份效应所需的信息是瓦尔德估计量的分母,也即应征资格对在军队服役概率的影响。表4.3的第(4)列报告了这一信息,其显示了符合应征资格的男性在越南战争时期服役的可能性要高出近16个百分点。第(5)列给出了1981年服役对收入影响的瓦尔德估计值,约占平均值的15%。1971年的效应更大(按百分比计算),当时受影响的士兵仍在军队中服役。

瓦尔德/工具变量估计量的一个重要特征是,这些识别假设很容易评估和解释。$\mathbf{D}_i$表示军人身份,$\mathrm{Z}_i$表示有无应征资格。证明我们将**瓦尔德估计量解释为体现$\mathbf{D}_i$的因果效应的基本主张是$E\left[\mathrm{Y}_i \mid \mathrm{Z}_i\right]$随$\mathrm{Z}_i$变化而变化的唯一原因是$E\left[\mathrm{D}_i \mid \mathrm{Z}_i\right]$的变化。**对这一点的一个简单检验是寻找$\mathrm{Z}_i$和个人特征之间的关联,而这些特征不应该受到$\mathbf{D}_i$的影响,例如种族、性别或任何其他在$\mathbf{D}_i$被确定之前就已经确定的特征。另一个有用的检验是在$\mathbf{D}_i$和$\mathrm{Z}_i$之间不存在关系的样本中寻找工具变量和结果之间的联系。**如果应征资格对收人产生影响的唯一原因是军人身份,那么在应征资格与军人身份无关的样本中,应征资格对收入的影响应该为零。**

这一观点在
Angrist(1990)通过观察1969年的收入(在表的最后一行重新给出的一项估计值)对应征资格抽签的研究中得到了阐述。令人欣慰的是,**应征资格对1969年收入的影响是零,因为
1969 年的收入早于1970年的应征抽签。**这个观点的第二个检验变式研究了 1953

RSN,但1953年出生的人实际上并没有被征召(征召正式结束于1973年7月)。因此,1953年出生的男性(使用
1952年的95

我们用一组关于家庭规模对母亲就业和工作的影响的工具变量估计值来结束对瓦尔德估计量的讨论。像学校教育和服兵役研究一样,这些估计值在本书的其他地方也会被用作阐释说明。生育和劳动力供给之间的关系一直是劳动经济学家感兴趣的问题,该问题中出现遗漏变量偏差的情况是很明显的;劳动力参与弱或收入潜力低的母亲或许比劳动力参与强或收入潜力高的母亲更有可能生孩子。这使得所观察到的家庭规模和就业之间的关联难以解释,因为有着大家庭的母亲可能相对工作得更少。Angrist和Evans(1998)使用两个工具变量解决了这个遗漏变量问题,这两个工具变量都借助了瓦尔德型估计策略。

第一个瓦尔德估计量使用的是多胎生育,这是一种由 Rosenzweig 和
Wolpin(1980)首创的家庭规模效应的识别策略。Angrist和Evans(1998)的双胞胎工具变量是多胎生育的一个虚拟变量,样本中的母亲都至少有两个孩子。双胞胎工具变量回归中第一阶段的估计值是0.625,表的第(3)列报告了这个估计值。这意味着,有两个或两个以上孩子的母亲中,37.5%无论如何都会生第三胎;多胎生育会使这一比例增加到1。双胞胎工具变量乃是基于这样一种观点,即多胎生育的发生基本上是随机的,与潜在的结果或家庭背景无关。

表中的第二个瓦尔德估计量使用了兄弟姐妹性别构成这一工具变量,这样做的动机是,已有两个孩子的美国父母,如果两个孩子是同性别而不是异性别,则他们更有可能有第三个孩子。这在表4.4的第(5)列中得到了说明,该列显示,有同性兄弟姐妹的父母生第三胎的可能性要高出6.7个百分点(有异性兄弟姐妹的父母生第三胎的可能性是0.38)。同性别工具变量的依据是,兄弟姐妹的性别构成基本上是随机的,仅通过提高生育率来影响家庭劳动力供给。

125,而1952年出生的男性的应征资格上限为RSN95。然而,实际上,抽签号(我们称为R,即RSN的缩写)和军人身份($\mathbf{D}_i$)之间的联系比应征资格身份本身更丰富多样。**虽然抽签值超过资格上限的男性并未被征召,但上限是多少事先是未知的。因此,有些人自愿参军,以期在更好的条件下服役,并获得对服役时间的一些控制权。对于抽签号码较小的男性来说,自愿服役的驱动力很大,但对于抽签号码较大的男性来说,这种驱动力是很小的。因此,即使抽签值严格高于或低于征召资格临界点,$P\left[\mathrm{D}_i=1 \mid \mathrm{R}_i\right]$也会有所变化。例如,1950年出生、抽签号码为200-225的男性比抽签号码为226—250的男性更有可能参军,尽管最终两组都没有受到征召。**

瓦尔德估计量将征召资格作为1950年出生的男性的一个工具变量,将$\mathrm{R}_i<195$的男性的收入与$\mathrm{R}_i>195$的男性的收入进行比较。但是前文的讨论表明了更多可能的比较,例如$\mathrm{R}_i \leq 25$的男性与$\mathrm{R}_i \in[26-50]$的男性的比较,$\mathrm{R}_i \in[51-75]$的男性与$\mathrm{R}_i \in[76-100]$的男性的比较,等等,直到这些25个数字的间隔被比较结束。**我们也可以让间隔更小一些,比如说,用5个数字或个位数的间隔来比较男性,而不是25个数字的间隔。这种扩大比较的结果是得到一组瓦尔德估计量。**这些集合是完全的,因为这些间隔划分了基础工具变量的支撑集合,而单个瓦尔德估计量之间是线性独立的,因为它们的分子是线性独立的。最后,只要$\mathrm{R}_i$独立于潜在结果,并与军人身份相关(即瓦尔德估计量的分母不为零),那么,这些瓦尔德估计量中的每一个都—致地估计了相同的因果效应,这里假设它为常数。

为相同的因果效应构建多个瓦尔德估计量的可能性,自然会引发如何处理所有这些估计量的问题。我们会想要给出一个单一的估计值,从而以某种方式有效地结合各个瓦尔德估计值中的信息。事实证明,一组线性独立的瓦尔德估计量的最有效的线性组合,是通过构建这些估计值的组平均值拟合一条直线产生的。

使用二元常量因果效应模型

$$
\mathrm{Y}_i=\alpha+\rho \mathrm{D}_i+\eta_i
$$

其中$\rho=\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i}$为我们感兴趣的因果效应,且$\mathrm{Y}_{0 i}=\alpha+\eta_i$。由于$\mathbf{R}_i$是随机分配的,而且假定抽签号除了通过军人身份对收入产生影响之外别无它途,因此$E\left[\eta_i \mid \mathrm{R}_i\right]=0$。进一步可以推断:

$$
E\left[\mathrm{Y}_i \mid \mathrm{R}_i\right]=\alpha+\rho P\left[\mathrm{D}_i=1 \mid \mathrm{R}_i\right]
$$

这是因为$P\left[\mathrm{D}_i=1 \mid \mathrm{R}_i\right]=E\left[\mathrm{D}_i \mid \mathrm{R}_i\right]$。换句话说,将给定抽签号码时的平均收入与按抽签号码服兵役的平均概率联系起来的直线的斜率,等于服兵役所产生的影响。而$\mathbf{Y}_i$对$\mathrm{D}_i$的回归结果——在这种情况下,根据军人身份得到的收入平均值的差异——几乎肯定异于的事实,乃是因为$\mathrm{Y}_{0 i}$和$\mathrm{D}_i$很可能相关。

**在
Angrist(1990)中,抽签号码信息被聚合到69个5个数字的间隔中,加上数字间隔346-365,一共70个间隔。因此,我们可以认为R,从1到70取值。令$\bar{y}_j$和$\hat{p}_j$表示$E\left[\mathrm{Y}_i \mid \mathrm{R}_i=j\right]$和$P\left[\mathrm{D}_i=1 \mid \mathrm{R}_i=j\right]$的估计值,而$\bar{\eta}_j$为上述表达式中的平均误差。由于样本矩收敛于总体矩,因此可以得出以下分组方程中的OLS估计值是一致的**:

$$
\bar{y}_j=\alpha+\rho \hat{p}_j+\bar{\eta}_j
$$

在实践中,GLS可能是更可取的,因为分组方程具有方差结构的已知异方差性。因此,我们应该按组的大小来衡量权重。

它的重要性在于。首先,由J组观测值构造的GLS斜率估计值是任意(J-1)个线性独立瓦尔德估计量的全集的渐近有效线性组合(Angrist,1991)。其次，正如每个瓦尔德估计量也是一个工具变量估计量一样，上式的GLS估计量也是2SLS的。

分组数据和 2SLS
之间的联系在概念上和实践上都具有重要意义。**在概念上,任何使用一组虚拟工具变量的
2SLS
估计量都可以被理解为依次使用这些工具变量产生的所有瓦尔德估计量的一个线性组合。**

在实践方面,2SLS的分组数据等价关系为我们提供了一个简单的工具,可以用来解释和评估任何工具变量策略。例如,在应征抽签的情况下,分组模型体现了这样一个假设:平均收入随抽签号码的变化而变化的唯一原因是不同抽签号码组服兵役的概率的变化。如果潜在的因果关系是线性的,且是常量因果效应,那么上式应能很好地拟合组平均值,这是我们可以通过检验来评估的,在下一节,我们还会讨论如何使用正式的统计推断机制来加以评估。

有时,劳动经济学家将离散工具变量的分组数据图称为可视化工具变量(visual
in-strumental variables, VIV),
Angrist(1990)中就有这样一个例子,如图4.2所重制。这张图显示了出生在1950-1953年的白人男性1981-1984年的收入中,以5个RSN数为间隔的组别平均收入与这些组别中服兵役的概率之间的关系。通过这些点的直线斜率是对因服兵役而造成的收入损失的工具变量估计值,在此例中约为2400美元,与之前讨论的瓦尔德估计值相差不大,但此例中的标准误更低(约为800美元)。

### 2SLS的渐近推断

它的渐近协方差矩阵为

$$
\left[\sum_i V_i V_i^{\prime}\right]^{-1}\left[\sum_i V_i V_i^{\prime} \eta_i^2\right]\left[\sum_i V_i V_i^{\prime}\right]^{-1}
$$

其中$V_i \equiv\left[\mathrm{X}_i^{\prime} \hat{s}_i\right]^{\prime}$

给定协方差，它可以简化为

$$
\left[\sum_i V_i V_i^{\prime}\right]^{-1} \sigma_\eta^2
$$

## 具有异质性潜在结果的工具变量

- **内部效度(internal validity)** -
  事关一个给定的设计是否成功地揭示了被研究总体的因果效应的问题。一项随机临床试验,或者说一项好的工具变量试验,有很强的内部效度。

- **外部效度(external validity)** -
  事关研究结果在不同背景下的预测价值。例如,如果随机试验中的研究总体极有可能从处理中受益,则得出的估计结果可能几乎没有外部效度。类似地,对越南战争时期征兵效应的抽签估计值不一定能很好地衡量自愿服兵役的因果影响。一个具有异质处理效应的计量经济学框架有助于我们评估工具变量估计值的內外部效度。

### 局部平均处理效应

我们可以把工具变量想象成一个因果链的引擎,工具变量$\mathrm{Z}_i$,影响我们感兴趣的变量$\mathbf{D}_i$,
$\mathbf{D}_i$反过来影响结果变量$\mathrm{Y}_i$。为了清晰起见,我们引入记号来表达工具变量对$\mathbf{D}_i$有因果效应的观点。设$\mathrm{D}_{1 i}$为$\mathrm{Z}_i=1$时$i$的处理状态,$\mathrm{D}_{0 i}$为$\mathrm{z}_i=0$时$i$的处理状态。因此观察到的处理状态为:

$$
\mathrm{D}_i=\mathrm{D}_{0 i}+\left(\mathrm{D}_{1 i}-\mathrm{D}_{0 i}\right) \mathrm{Z}_i=\pi_0+\pi_{1 i} \mathrm{Z}_i+\xi_i
$$

其中$\pi_0 \equiv E\left[\mathrm{D}_{0 i}\right]$和$\pi_{1 i} \equiv \left(\mathrm{D}_{1 i}-\mathrm{D}_{0 i}\right)$。所以是工具变量对$\mathbf{D}_i$的异质性因果效应。与潜在结果一样,对于任何一个人,我们只能观察到一种潜在的处理分配,即$\mathrm{D}_{1 i}$和$\mathbf{D}_{0 i}$。在抽签的例子中,$\mathbf{D}_{0 i}$指的是如果$i$抽到一个大的(不合格的)抽签号码,他是否会服兵役,而$\mathrm{D}_{1 i}$指的是如果$i$抽到一个小的(合格的)抽签号码,他是否会服兵役。**我们可以看到这些一个又一个的潜在赋值取决于$\mathrm{Z}_i$。$\mathrm{Z}_i$对$\mathrm{D}_i$的平均因果效应为$E\left[\pi_{1 i}\right]$。**

异质效应框架中的第一个假设是,**该工具与随机分配效果一样,它与潜在结果向量和潜在处理分配无关。**正式地,这可以写成

$$
\left[\left\{\mathrm{Y}_i(d, z) ; \forall d, z\right\}, \mathrm{D}_{1 i}, \mathrm{D}_{0 i}\right] \perp \mathrm{z}_i
$$

这个独立性假设足以解释简化形式的因果关系,即$Y_i$对$\mathbf{Z}_i$的回归。

$$
\begin{aligned}
& E\left[\mathrm{Y}_i \mid \mathrm{Z}_i=1\right]-E\left[\mathrm{Y}_i \mid \mathrm{Z}_i=0\right] \\
& \quad=E\left[\mathrm{Y}_i\left(\mathrm{D}_{1 i}, 1\right) \mid \mathrm{z}_i=1\right]-E\left[\mathrm{Y}_i\left(\mathrm{D}_{0 i}, 0\right) \mid \mathrm{z}_i=0\right] \\
& \quad=E\left[\mathrm{Y}_i\left(\mathrm{D}_{1 i}, 1\right)-\mathrm{Y}_i\left(\mathrm{D}_{0 i}, 0\right)\right]
\end{aligned}
$$

这就是工具变量对$\mathrm{Y}_i$的因果效应。独立性也意味着:

$$
\begin{aligned}
E\left[\mathrm{D}_i \mid \mathrm{z}_i=1\right]-E\left[\mathrm{D}_i \mid \mathrm{z}_i=0\right] & =E\left[\mathrm{D}_{1 i} \mid \mathrm{z}_i=1\right]-E\left[\mathrm{D}_{0 i} \mid \mathrm{z}_i=0\right] \\
& =E\left[\mathrm{D}_{1 i}-\mathrm{D}_{0 i}\right] 
\end{aligned}
$$

异质效应框架中的**第二个关键假设是$\mathrm{Y}_i(d, z)$仅是$d$的函数。**具体而言,虽然应征资格明显影响退伍军人身份,但假设个人作为退伍军人或非退伍军人的潜在收入不会因应征资格而改变。一般来说,**工具变量通过已知的单一因果路径作用的主张被称为排他性约束。**正式地说,排他性约束表示如下:

$$
\mathrm{Y}_i(d, 0)=\mathrm{Y}_i(d, 1) \text { 对于 } d=0,1
$$

如果征兵抽签号码较小的男性,除服兵役的可能性增加外,还受到其他(因素)的影响,那么排他性约束就不成立了。例如,Angrist
和Kruger
(1992)寻找抽签号码和学校教育之间的联系。他们的想法是,**接受教育可暂缓服兵役,这可能会让那些抽签号码较小的人在大学里待的时间比他们原本希望的更长。如果是这样,那么抽签号码与收入相关至少有两个原因:服兵役的可能性增加和上大学的可能性增加。**抽签号码是随机分配的(因此满足独立性假设)这一事实并不会降低这种可能性。排他性约束就不同于工具变量和随机分配效果一样的主张了。更确切地说,它是一种关于工具变量因果效应的唯一渠道的主张。

对于异质性工具变量模型,最后一个假设是对于所有$i$,$\pi_{1 i} \geq 0$或$\pi_{1 i} \leq 0$,两者只择一。Imbens
和 Angrist(1994)提出的这个**单调性假设(monotonicity
assumption)**意味着,**虽然工具变量可能对某些人没有影响,但所有受影响的人都以相同的方式受到影响。**接句话说,对于所有$i$,要么$\mathrm{D}_{1 i} \geq \mathrm{D}_{0 i}$,要么$\mathrm{D}_{1 i} \leq \mathrm{D}_{0 i}$。接下来,我们假设当$\mathrm{D}_{1 i} \geq \mathrm{D}_{0 i}$时单调性成立。在抽签征兵的例子中,这意味着尽管征兵资格可能对某些人服兵役的概率没有影响,但没有人因为有征兵资格却实际上被排除在军队之外。如果没有单调性,工具变量估计量就不能保证估计出潜在的个体因果效应的加权平均值$\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i}$。

基于排他性约束、工具变量和潜在结果的独立性、第一阶段的存在以及单调性这一系列假设,瓦尔德被估量可以被解释为退伍军人身份对那些可以由工具变量改变处理状态的人的影响。该参数被称为局部平均处理效应(local
average treatment effect, LATE;Imbens and Angrist,1994)。

**定理 4.4.1 LATE定理**

假设

- A1（独立性）-
  $\left\{\mathrm{Y}_i\left(\mathrm{D}_{1 i}, 1\right), \mathrm{Y}_i\left(\mathrm{D}_{0 i}, 0\right), \mathrm{D}_{1 i}, \mathrm{D}_{0 i}\right\} \perp \mathrm{Z}_i$

- A2（排他性）-
  $\mathrm{Y}_i(d, 0)=\mathrm{Y}_i(d, 1) \equiv \mathrm{Y}_{d i}$ for
  $d=0,1$

- A3（第一阶段）-
  $E\left[\mathrm{D}_{1 i}-\mathrm{D}_{0 i}\right] \neq 0$

- A4（单调性）- $\mathrm{D}_{1 i}-\mathrm{D}_{0 i} \geq 0 \forall i$, or
  vice versa

那么有

$$
\begin{aligned}
\frac{E\left[\mathrm{Y}_i \mid \mathrm{Z}_i=1\right]-E\left[\mathrm{Y}_i \mid \mathrm{Z}_i=0\right]}{E\left[\mathrm{D}_i \mid \mathrm{Z}_i=1\right]-E\left[\mathrm{D}_i \mid \mathrm{Z}_i=0\right]} & =E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_{1 i}>\mathrm{D}_{0 i}\right] \\
& =E\left[\rho_i \mid \pi_{1 i}>0\right]
\end{aligned}
$$

### 依从性子总体

LATE
框架使用某一工具变量将任一总体划分为三个与工具变量相关的子总体,根据它们对该工具变量的反应方式定义。

**定义4.4.1**

- 依从者:$\mathrm{D}_{1 i}=1$和$\mathrm{D}_{0 i}=0$的子总体。

- 总是依从者:$\mathrm{D}_{1 i}=\mathrm{D}_{0 i}=1$的子总体。

- 从不依从者:$\mathrm{D}_{1 i}=\mathrm{D}_{0 i}=0$的子总体。

LATE
是指对依从者总体的处理效应。术语“依从者”来自对随机试验的类比。在随机试验中,有些实验对象遵守随机分配的治疗方案(例如,服药),但也有一些人不遵守,还有一些对照对象却获得了实验治疗,即使他们原本不应该获得。被随机分配到不服药组的人为从不依从者(从不服药的人),而被分配到对照组的人是总是依从者(即总是服药的人)。如果不添加进一步的假设(例如,常量因果效应),LATE就不能提供关于对从不服药的人(从不依从者)和总是服药的人(总是依从者)的影响的信息,因为根据定义,这两组人的处理状态不会因工具变量而改变。工具变量与带有部分依从性的随机试验之间的类比不仅仅是个比喻:工具变量解决了带有部分依从性的随机试验中的因果推断问题。

**一些观点**

- 对依从者的平均因果效应通常不等同于对处理组的平均因果效应。

$$
\begin{aligned}
& \underbrace{E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_i=1\right]}_{\text {接受处理的效应}} \\
& \quad=\underbrace{E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_{0 i}=1\right]}_{\text {总是依从者的效应 }} P\left[\mathrm{D}_{0 i}=1 \mid \mathrm{D}_i=1\right] \\
& \quad+\underbrace{E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_{1 i}>\mathrm{D}_{0 i}\right]}_{\text {依从接收者的效应 }} P\left[\mathrm{D}_{1 i}>\mathrm{D}_{0 i}, \mathrm{Z}_i=1 \mid \mathrm{D}_i=1\right]
\end{aligned}
$$

- LATE也不等于处理对未受处理者的平均因果效应$E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_i=0\right]$

$$
\begin{aligned}
& \underbrace{E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_i=0\right]}_{\text {未受处理的效应 }} \\
& \quad=\underbrace{E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_{1 i}=0\right]}_{\text {从不接收者的效应 }} P\left[\mathrm{D}_{1 i}=0 \mid \mathbf{D}_i=0\right] \\
& \quad+\underbrace{E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathbf{D}_{1 i}>\mathrm{D}_{0 i}\right]}_{\text {依从接受者的效应 }} P\left[\mathrm{D}_{1 i}>\mathrm{D}_{0 i}, \mathrm{Z}_i=0 \mid \mathbf{D}_i=0\right]
\end{aligned}
$$

最后

$$
\begin{aligned}
E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i}\right]= & E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_i=1\right] P\left[\mathrm{D}_i=1\right] \\
& +E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_i=0\right] P\left[\mathrm{D}_i=0\right]
\end{aligned}
$$

这一结果显示,无条件平均处理效应是对依从者、总是依从者和从不依从者的加权平均效应。

### 随机试验中的工具变量

LATE
框架的语言是基于工具变量和随机试验之间的类比给出的。而有些工具变量确实来自随机试验。如果工具变量得自处理的一个随机分配要求,那么LATE
指的就是对依从要求却未被处理的那些人的处理效应。一个特别重要的情况就是,当工具变量由一项单边不依从(one-sided
noncompliance)的随机试验产生。在许多随机试验中,随机分配处理的人是自愿参与实验的。另一方面,对照组的人没有机会接受实验处理。由于接受(随机)分配处理的组是那些被提供了处理的自选择个体,所以在实际的处理组和对照组之间的比较是具有误导性的。这种情况下的选择性偏差几乎总是为正的:那些在随机试验中服用药物的人往往更健康;那些利用随机分配的经济干预措施(如培训项目)的人不管怎样往往都获得了更高的回报。

使用随机分配处理的工具变量作为接受处理的工具变量,解决了这类依从性问题。此外,在这种情况下,LATE是对处理组的处理效应。**假设工具变量$\mathrm{Z}_i$,是一个虚拟变量,表示对处理组的随机分配,而$\mathbf{D}_i$是一个虚拟变量,表示是否实际接受处理。在实践中,由于不依从性的存在$\mathbf{D}_i$不等于$\mathrm{Z}_i$。一个例子是对JTPA培训项目的随机评估,其中只有46.0%的被分配到培训组的人接受了培训,而大约2%的被分配到对照组的人总是接受了培训训(Bloom
et al.,
1997)。不依从JTPA的原因是参与者缺乏兴趣,以及项目运营者未能鼓励参与者参与。由于该案例的依从性问题很大程度上局限在处理组,所以使用(表示)随机分配的$\mathrm{Z}_i$作为接受处理的工具变量$\mathrm{D}_i$的LATE,是对处理组的处理效应。**

**定理 4.4.2**

布鲁姆的结论。假定LATE定理的假设都成立,而且$E\left[\mathrm{D}_i \mid \mathrm{z}_i=0\right]=P\left[\mathrm{D}_i=1 \mid \mathrm{z}_i=0\right]=0$。那么有

$$
\frac{E\left[\mathrm{Y}_i \mid \mathrm{Z}_i=1\right]-E\left[\mathrm{Y}_i \mid \mathrm{Z}_i=0\right]}{P\left[\mathrm{D}_i=1 \mid \mathrm{Z}_i=1\right]}=E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_i=1\right]
$$

除了告诉我们如何分析具有不依从性的随机试验,LATE框架还打开了一扇门:在强迫依从处理不可能也不道德的环境中,如何巧妙地设计随机试验。犯罪学领域的一个有创意的例子是明尼阿波利斯家庭暴力实验(MDVE)。MDVE做出了确定警察应对家庭暴力的最佳对策的开创性努力(Sherman
and Berk,
1984),一般来说,警方在接到家庭暴力报警电话时会采取多种策略,包括问询、分居令和逮捕。一场激烈的辩论围绕着强硬的举措————逮捕或者至少暂时监禁——————是否有效展开,尤其是考虑到家暴指控经常被撤销的事实。

这场辩论的结果是,明尼阿波利斯市授权进行了一项随机实验,在这项实验中,警方对家庭暴力报警的反应部分地通过随机分配来决定。研究设计使用随机洗牌(得到)的顏色编号报告表,来告诉执行警官是对某些人实施逮捕,或是对某些人进行问询,或是仅下达分居令。然而,在实践中,警察可以自由地否决这种随机分配。例如,一个特别危险的或醉酒的嫌犯无论如何都会被逮捕。因此,实际的反应往往偏离随机分配的反应,尽管这两者是高度相关的。

大多数已发表的 MDVE
数据分析都认识到这一依从性问题,并关注ITT效应,也就是说,他们使用的是原始的随机分配,而不是实际提供的处理。

## LATE的扩展

LATE
定理适用于一个简化的因果模型,在该模型中只有单个虚拟工具被用来估计没有协变量的虚拟处理的因果效应。我们可以通过三种重要的方式来扩展这一点:多个工具变量的模型(例如,一组出生季度的虚拟变量),带有协变量的模型(例如,对出生年份的控制),以及工具变量具有可变和连续处理强度的模型(例如,受教育年限)。在所有三种情况下,工具变量被估量是对特定工具变量依从者的因果效应的加权平均。

### 具有多个工具变量的LATE

多工具变量的扩展是很容易理解的。这与我们在分组数据情形中讨论的结果基本相同。考虑一对虚拟工具变量和22。在不失一般性的前提下,假设这些虚拟变量是互斥的,,那么这两个虚拟变量可以被用来构造瓦尔德估计量。同样,在不失一般性的前提下,假设每个的第一阶段的单调性都是满足的且为正值.因此,尽管满足$\mathbf{D}_{1 i}>\mathbf{D}_{0 i}$的总体在$\mathbf{Z}_{1 i}$和$\mathbf{Z}_{2 i}$上是不同的,但两个估计都是$E\left[\mathrm{Y}_{1 i}-\mathrm{Y}_{0 i} \mid \mathrm{D}_{1 i}>\mathrm{D}_{0 i}\right]$的类似形式。

### 异质性效应模型中的协变量

在使用工具变量的因果分析中纳入协变量的主要原因是:在对协变量进行条件约束后,工具变量估计基础上的条件独立性和排他性约束可能更有效。即使是随机分配的工具变量,比如应征资格状态,也可能只在有协变量条件的约束下才有效。在应征资格方面,年龄较大的人群更有可能应征,因为应征资格的临界点更高。由于在收入上存在出生年份(或年龄)的差异,所以应征资格只在以出生年份为条件约束下才是有效的工具变量。

更正式点说,带有协变量的工具变量估计可以用条件独立性假设来给出合理性证明:

$$
\left\{\mathrm{Y}_{1 i}, \mathrm{Y}_{0 i}, \mathrm{D}_{1 i}, \mathrm{D}_{0 i}\right\} \perp \mathrm{Z}_i \mid \mathrm{X}_i
$$

换言之,**在以协变量$\mathrm{X}_i$为条件后(这里我们仍然隐含地保留排他性约束),我们把工具变量看成“与随机分配一样好”。引入协变量的第二个原因是,以协变量为条件约束可以减少因变量的一些可变性。这可以推导出更精确的2SLS估计值。**
