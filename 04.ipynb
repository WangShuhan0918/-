## 平行世界：固定效应、双重差分法和面板数据

## 个体固定效应

- 问题：工人们集体谈判下的工资会因工会成员（身份）而增加吗？

令$\mathrm{Y}_{i t}$等于工人$i$在时间$t$时的（对数）收入，令$\mathrm{D}_{i t}$表示他是否为工会成员。假设

$$
E\left[\mathrm{Y}_{0 i t} \mid A_i, \mathrm{X}_{i t}, t, \mathrm{D}_{i t}\right]=E\left[\mathrm{Y}_{0 i t} \mid A_i, \mathrm{X}_{i t}, t\right]
$$

其中$\mathrm{X}_{i t}$是一个协变量向量，$A_i$是一个未观察到的混杂因素向量，例如能力。因此

$$
E\left[\mathrm{Y}_{0 i t} \mid A_i, \mathrm{X}_{i t}, t\right]=\alpha+\lambda_t+A_i^{\prime} \gamma+\mathrm{X}_{i t}^{\prime} \beta
$$

假设工会成员的因果效应是可加的，并且是一个常量

$$
E\left[\mathrm{Y}_{1 i t} \mid A_i, \mathrm{X}_{i t}, t\right]=E\left[\mathrm{Y}_{0 i t} \mid A_i, \mathrm{X}_{i t}, t\right]+\rho
$$

结合在一起，得到

$$
E\left[\mathrm{Y}_{i t} \mid A_i, \mathrm{X}_{i t}, t, \mathrm{D}_{i t}\right]=\alpha+\lambda_t+\rho \mathrm{D}_{i t}+A_i^{\prime} \gamma+\mathrm{X}_{i t}^{\prime} \beta
$$

它意味着

$$
\mathrm{Y}_{i t}=\alpha_i+\lambda_t+\rho \mathrm{D}_{i t}+\mathrm{X}_{i t}^{\prime} \beta+\varepsilon_{i t}
$$

且有

$$
\alpha_i \equiv \alpha+A_i^{\prime} \gamma
$$

这是一个固定效应模型

**固定效应估计量**

首先，计算个体的平均值

$$
\overline{\mathrm{Y}}_i=\alpha_i+\bar{\lambda}+\rho \overline{\mathrm{D}}_i+\overline{\mathrm{X}}_i^{\prime} \beta+\bar{\varepsilon}_i
$$

除时间均值

$$
\mathrm{Y}_{i t}-\overline{\mathrm{Y}}_i=\lambda_t-\bar{\lambda}+\rho\left(\mathrm{D}_{i t}-\overline{\mathrm{D}}_i\right)+\left(\mathrm{X}_{i t}-\overline{\mathrm{X}}_i\right)^{\prime} \beta+\left(\varepsilon_{i t}-\bar{\varepsilon}_i\right)
$$

这抵消了未观测的个体效应。

也可以使用差分方法

$$
\Delta \mathrm{Y}_{i t}=\Delta \lambda_t+\rho \Delta \mathrm{D}_{i t}+\Delta \mathrm{X}_{i t}^{\prime} \beta+\Delta \varepsilon_{i t}
$$

## 倍差法（双重差分法）：事前与事后，处理与控制

假设我们对最低工资如何影响就业感兴趣,这是劳动经济学中的一个经典问题。在一个竞争性的劳动力市场中,最低工资的提高使向右下方倾斜的劳动力需求曲线向上移动。因此,提高最低工资会减少就业,这可能会伤害到那些最低工资政策本想帮助的工人。Card
和 Krueger(1994)利用新泽西州最低工资的突发变化来验证这一观点是否正确。

1992年4月1日,新泽西州将州最低小时工资从4.25美元提高到5.05美元。Card
和Krueger(1994)在1992年2月和11月两次收集了新泽西州快餐店的就业数据。这些餐厅(汉堡王、温蒂汉堡等)的员工都拿最低工资。他们还从特拉华河对岸的宾夕法尼亚州东部同类型餐厅收集了数据。在此期间,宾夕法尼亚州的最低小时工资一直稳定在4.25美元。他们使用这些数据集来计算提高新泽西州最低工资的影响的DID估计。也就是说,他们将新泽西州2月至11月的就业变化与宾夕法尼亚州同期的就业变化进行了比较。

DID
是使用加总数据进行固定效应估计的一种做法。为了弄清楚这一点,如果州最低小时工资较高,则设快餐店就业情况为$\mathbf{Y}_{1 i s t}$($s$表示州,$i$表示快餐店,$t$表示时期);如果州最低小时工资较低,则设快餐店就业情况为$\mathrm{Y}_{0 i s t}$(s表示州,i表示快餐店,t表示时期)。这些都是潜在的结果;实际上,我们只能看到其中之一。例如,我们只能看到1992年11月新泽西州的。DID框架的核心是无处理状态下潜在结果的可加性结构。具体来说,我们假设:

$$
E\left[\mathrm{Y}_{0 i s t} \mid s, t\right]=\gamma_s+\lambda_t
$$

这个等式表明,在最低工资无变化的状态下,就业是由不随时期变化的州效应和不随州变化的年份效应的总和决定的。

令$\mathbf{D}_{s t}$表示时期$t$最低工资较高的州$s$的虚拟变量。设$E\left[\mathrm{Y}_{1 i s t}-\mathrm{Y}_{0 i s t} \mid s, t\right]$为一个常量，记为$\delta$。因此

$$
\mathbf{Y}_{i s t}=\gamma_s+\lambda_t+\delta \mathbf{D}_{s t}+\varepsilon_{i s t}
$$

由此可得

$$
\begin{aligned}
& E\left[\mathrm{Y}_{i s t} \mid s=P A, t=N o v\right]-E\left[\mathrm{Y}_{i s t} \mid s=P A, t=F e b\right] \\
& \quad=\lambda_{N o v}-\lambda_{\text {Feb }}
\end{aligned}
$$

和

$$
\begin{aligned}
& E\left[\mathrm{Y}_{i s t} \mid s=N J, t=N o v\right]-E\left[\mathrm{Y}_{i s t} \mid s=N J, t=F e b\right] \\
& \quad=\lambda_{N o v}-\lambda_{F e b}+\delta 
\end{aligned}
$$

下述总体DID就是我们感兴趣的因果效应

$$
\begin{aligned}
& \left\{E\left[\mathrm{Y}_{i s t} \mid s=N J, t=N o v\right]-E\left[\mathrm{Y}_{i s t} \mid s=N J, t=F e b\right]\right\} \\
& \quad-\left\{E\left[\mathrm{Y}_{i s t} \mid s=P A, t=N o v\right]-E\left[\mathrm{Y}_{i s t} \mid s=P A, t=F e b\right]\right\}=\delta
\end{aligned}
$$

### DID回归

令$N J_s$为新泽西州餐馆的虚拟变量，$d_t$为11月（即最低小时工资改变后）观测结果的时间虚拟变量。那么

$$
\mathrm{Y}_{i s t}=\alpha+\gamma N J_s+\lambda d_t+\delta\left(N J_s \cdot d_t\right)+\varepsilon_{i s t}
$$

$$
\begin{aligned}
\alpha= & E\left[\mathrm{Y}_{i s t} \mid s=P A, t=F e b\right]=\gamma_{P A}+\lambda_{F e b} \\
\gamma= & E\left[\mathrm{Y}_{i s t} \mid s=N J, t=F e b\right]-E\left[\mathrm{Y}_{i s t} \mid s=P A, t=F e b\right] \\
= & \gamma_{N J}-\gamma_{P A} \\
\lambda= & E\left[\mathrm{Y}_{i s t} \mid s=P A, t=N o v\right]-E\left[\mathrm{Y}_{i s t} \mid s=P A, t=F e b\right] \\
= & \lambda_{N o v}-\lambda_{F e b} \\
\delta= & \left\{E\left[\mathrm{Y}_{i s t} \mid s=N J, t=N o v\right]-E\left[\mathrm{Y}_{i s t} \mid s=N J, t=F e b\right]\right\} \\
& -\left\{E\left[\mathrm{Y}_{i s t} \mid s=P A, t=N o v\right]-E\left[\mathrm{Y}_{i s t} \mid s=P A, t=F e b\right]\right\}
\end{aligned}
$$

DID 模型的回归公式为构建DID估计值和标准误提供了一种便利的方法。

回归 DID
的第二个优点是,它不只是便利了那些可以用虚拟变量描述的政策,更是便利了对一般性政策的研究。例如,我们可以研究美国所有州的最低工资标准,而不是
1992年的新泽西州和宾夕法尼亚州。其中一些略高于联邦最低标准(联邦最低标准涵盖所有人,无论他们住在哪里),一些高得多,还有一些则是一样的。因此,最低工资是一个变量,不同州和不同时期的处理强度不同。此外,除了受州最低工资的立法变化影响外,当地最低工资的重要性还会随着州平均工资水平的变化而变化。例如,20世纪90年代初,联邦最低时薪4.25美元在平均小时工资较高的康涅狄格州可能无关紧要,但在密西西比州却极其重要。

Card(1992)探讨了联邦最低小时工资影响的地区差异。

$$
\mathbf{Y}_{i s t}=\gamma_s+\lambda_t+\delta\left(\mathrm{FA}_s \cdot d_t\right)+\varepsilon_{i s t}
$$

其中,变量$\mathrm{FA}_S$是每个州可能受到最低小时工资提高影响的青少年比例的度量指标,$d_t$是1990年观察值的一个虚拟变量,当时联邦最低小时工资从3.35美元增加到3.80美元。$\mathrm{FA}_S$变量衡量的是每个州每小时收入低于3.80美元的青少年劳动力的基线(增长前)比例。

在新泽西州——宾夕法尼亚的研究中,Card(1992)使用了两个时期的数据,分别是在最低工资立法变化之前的1989年和变化之后的1990年。报告的估计值来自以下一阶差分方程：

$$
\Delta \overline{\mathrm{Y}}_s=\lambda^*+\delta \mathrm{FA}_s+\Delta \bar{\varepsilon}_s
$$

可以在回归的框架中添加控制变量

$$
\mathrm{Y}_{i s t}=\gamma_s+\lambda_t+\delta\left(\mathrm{FA}_s \cdot d_t\right)+\mathrm{X}_{i s t}^{\prime} \beta+\varepsilon_{i s t}
$$

其中，$X_{ist}$可以包括个体层面的特征（如种族），以及在州层面观测到的随时间变化的变量。只有后者可能是遗漏变量偏误的来源，但个体层面的控制变量可以用来提高估计精度

当样本包含多个年份时，DID回归模型适用于格兰杰(Granger,
1969)思想下的因果关系检验。格兰杰的思想是看原因是否先于结果发生，而不是相反(尽管我们知道，只凭这点是不足以进行因果推断的)。假设感兴趣的政策变量$\mathrm{D}_{s t}$在不同的时间、不同的州发生变化。在这种情况下，格兰杰因果关系检验意味着检验在州和年份的影响下，过去的$\mathrm{D}_{s t}$是否能预测$\mathbf{Y}_{i s t}$，而未来的$\mathrm{D}_{s t}$是否无法预测$\mathbf{Y}_{i s t}$。

$$
\mathrm{Y}_{i s t}=\gamma_s+\lambda_t+\sum_{\tau=0}^m \delta_{-\tau} \mathrm{D}_{s, t-\tau}+\sum_{\tau=1}^q \delta_{+\tau} \mathrm{D}_{s, t+\tau}+\mathrm{X}_{i s t}^{\prime} \beta+\varepsilon_{i s t}
$$

Autor
(2003)在一项关于就业保护政策对企业雇用临时工的影响的研究中实施了格兰杰检验。在美国,就业保护政策是一种由州立法机构颁布的劳动法,或者更典型的是由州法院制定的普通法,这使得解雇工人更加困难。通常,美国劳动法允许企业随意雇佣,州法院制定的主的心血来潮,以正当理由或无理由的方式的诉讼法院允许“随意雇佣”原则存在一些例外情况,这就导致了应对不公正解雇的诉讼。

Autor(2003)感兴趣的是,对员工诉讼的担心是否会让公司更有可能雇用临时工来完成原本需要增加劳动力来完成的任务。临时工被公司以外的第三方雇用来完成任务。因此,雇用临时工的公司在解雇临时工时,不会因不正当解雇而被起诉。

Autor(2003)的经验策略将一个州的临时工雇用与虚拟变量联系起来,这些虚拟变量表明州法院的裁决允许随意雇佣原则的例外情况的存在。他的DID回归模型包括前导和滞后效应。从之前两年到之后四年的前导和滞后效应都被绘制在图中,这是Autor(2003)的图的再现。估计显示,在法院允许存在例外情况之前的两年里临时就业没有受到任何影响,而在允许存在例外情况后的头几年里,影响急剧增加,然后受影响州的临时就业率始终处于高位后而趋于平稳。这种模式似乎与Autor(2003)结果的因果解释是一致的。

### 挑选控制变量

我们标记了DID框架中州和时间两个维度,因为这是应用计量经济学中DID的典型示例。但是DID的思想要普遍得多。下标$s$可能表示人口统计**群组**,而不是州,群组中一些个人会受到政策的影响,而另一些个人则不会。例如,Kugler、Jimeno
和Hernanz(2005)研究了西班牙针对年龄的就业保护政策的效果。同样,我们也可以根据群组或其他类型的特征对数据进行分组,而不是按时间进行分组。例如
Angrist 和
Evans(1999),他们研究了各州堕胎法的变化对青少年怀孕的影响,使用的是各州和出生年份的变化。然而,不管分组标签是什么,DID设计总是构造一个隐含的处理组和对照组的比较。这种比较是否正确的问题值得仔细考虑。

在这种情况下,当处理组和对照组的构成因处理结果而改变时,就会出现一个潜在的陷阱。回到基于州和时间比较的设计中,假设我们感兴趣的是公共援助的慷慨程度对劳动力供给的影响。历史上,美国各州为贫穷的未婚母亲提供了各种各样的福利。长期以来,劳动经济学家一直对这种收入救济政策的效果很感兴趣:它们多大程度上促进了生活水平的提高,以及它们是否降低了工作的吸引力?\[例如,参见Meyer
和
Rosenbaum(2001)的一项最新研究。\]Moffitt(1992)在一篇关于福利研究的综述中强调,这里的一个问题是,穷人无论如何都有着较弱的劳动力依附度,因此他们可能会搬到福利待遇更优厚的州。在DID研究设计中,这种由项目引起的迁移往往使慷慨的福利项目对劳动力供给的影响看起来比实际情况更糟糕些。

如果我们知道个体的迁出地,迁移问题通常可以解决。比如说我们知道处理前的居住州,或者出生地所在州。出生地所在州或以前的居住州不会因处理而改变,但仍然与当前的居住州高度相关。因此,在使用这些维度而不是当前的居住州进行比较时,迁移问题就被消除了。

2×2的DID框架的修正对对照组进行了可能的改进,使用高阶对比来得出因果推论。一个例子是由Yelowitz(1995)研究的美国医疗补助覆盖范围的扩展。美国为穷人提供大范围的医疗保险计划,其中的医疗补助的获取资格,一度与获得抚养未成年子女家庭援助(AFDC)的资格挂钩,AFDC则是一个大型现金福利计划。然而,在20世纪80年代的各个时期,一些州将医疗补助覆盖范围扩大到不符合AFDC条件的家庭中的儿童。Yelowitz(1995)感兴趣的是,公共机构为儿童提供的医疗保险的扩大如何影响母亲的劳动参与率和收入等问题。除了州和时间,儿童的年龄提供了医疗补助政策变化的第三个维度。Yelowitz(1995)通过估计探究了这种变化:

$$
\mathbf{Y}_{\text {iast }}=\gamma_{s t}+\lambda_{a t}+\theta_{a s}+\delta \mathrm{D}_{\text {ast }}+\mathrm{X}_{\text {iast }}^{\prime} \beta+\varepsilon_{\text {iast }}
$$

## 固定效应与滞后因变量

固定效应和DID 估计量基于不随时间变化(或不随群组变化)遗漏变量的假设。例如

$$
E\left[\mathrm{Y}_{0 i t} \mid \alpha_i, \mathrm{X}_{i t}, \mathrm{D}_{i t}\right]=E\left[\mathrm{Y}_{0 i t} \mid \alpha_i, \mathrm{X}_{i t}\right]
$$

对于许多因果问题来说,最重要的遗漏变量不随时间变化这一设定似乎是不合理的。培训项目的评估就是一个很好的例子。那些希望通过参加政府资助的培训项目来改善劳动力市场选择的人很可能遭遇了某种挫折。许多培训项目明确针对最近遭遇挫折的人,比如最近失业的人。与此一致的是,Ashenfelter(1978)、Ashenfelter
和
Card(1985)发现,培训参与者的收入历史通常表现为参与项目前的下降。过去的收入是一个随时间变化的混杂变量,不能被包含在像$\alpha_i$这样的随时间不变的遗漏变量中。

受训人员不同的收入历史引发了一种估计策略,该策略直接控制过去的收入并忽略固定效应。准确地说,我们可以将因果推断建立在以下条件独立假设上

$$
E\left[\mathrm{Y}_{0 i t} \mid \mathrm{Y}_{i t-h}, \mathrm{X}_{i t}, \mathrm{D}_{i t}\right]=E\left[\mathrm{Y}_{0 i t} \mid \mathrm{Y}_{i t-h}, \mathrm{X}_{i t}\right]
$$

可以使用面板数据进行估计

$$
\mathbf{Y}_{i t}=\alpha+\theta \mathbf{Y}_{i t-b}+\lambda_t+\delta \mathrm{D}_{i t}+\mathrm{X}_{i t}^{\prime} \beta+\varepsilon_{i t}
$$

使用面板数据的应用研究者经常面临选择固定效应模型还是滞后因变量模型的挑战。解决这一困境的方法之一是,使用一个既包括滞后因变量又包括未观察到的个体效应的模型。换句话说,识别可能是基于:

$$
E\left[\mathrm{Y}_{0 i t} \mid \alpha_i, \mathrm{Y}_{i t-h}, \mathrm{X}_{i t}, \mathrm{D}_{i t}\right]=E\left[\mathrm{Y}_{0 i t} \mid \alpha_i, \mathrm{Y}_{i t-h}, \mathrm{X}_{i t}\right]
$$

我们可以尝试使用以下设定来估计因果效应

$$
\mathrm{Y}_{i t}=\alpha_i+\theta \mathrm{Y}_{i t-h}+\lambda_t+\delta \mathrm{D}_{i t}+\mathrm{X}_{i t}^{\prime} \beta+\varepsilon_{i t}
$$
