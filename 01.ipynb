## 因果与预测

为什么最佳的因果模型不一定最佳的预测模型呢？
这两个概念之间存在关联似乎是合理的：通常，导致其他事物的事物也会是预测因子。一切都归结于因果关系，因此任何预测信息都在某种程度上与我们正在预测的事物的因果结构相关。那么，一个预测效果好的模型也具有因果性，这难道不是顺理成章的吗？
的确，有些预测模型可以是优秀的因果模型，反之亦然。
遗憾的是，情况并非总是如此；因果效应不一定能很好地预测，而好的预测因子也不一定在因果上是无偏的（Shmueli
2010）。 仅凭数据无法知道这一点。

让我们首先从因果的角度来看，因为它更简单一些。
考虑一个因果上无偏的模型，该模型仅包含与结果和暴露相关的变量。
换句话说，这个模型为我们提供了关于感兴趣的暴露的正确答案，但不包括结果的其他预测因子。
如果一个结果有很多原因，那么一个准确描述与单一暴露关系的模型可能无法很好地预测结果。
同样，如果暴露对结果的真实因果效应很小，它也不会带来多少预测价值。
换句话说，模型预测能力的高低并不能帮助我们区分模型是否给出了正确的答案。
当然，较低的预测能力也可能表明，从应用的角度来看，因果效应并没有多大用处，尽管这取决于几个统计因素。

预测模型并不总是无偏因果模型还有两个更复杂的原因。
对于第一个原因，让我们考虑一个从因果角度来看准确的模型：它估计了对结果的影响，并且所有这些影响都是无偏的。
即使在这种理想的情况下，你也可以使用不同的模型获得更好的预测。
原因在于预测建模中的偏差-方差权衡。
当效应较小、数据嘈杂、预测因子高度相关或数据不多时，使用像惩罚回归这样的有偏模型可能是有意义的。
这些模型有意引入偏差，以提高数据外预测的方差。
由于预测和因果推断的目标不同（通常是为了数据外观察的准确预测，而不是无偏效应），因此用于推断的最佳模型不一定是最佳预测模型。

其次，从因果角度来看有偏的变量通常会带来预测能力。
让我们考虑一个简单的例子。
一个著名的混淆关系例子是夏季冰淇淋的销量和犯罪率。
从描述性角度来看，冰淇淋的销量和犯罪率是相关的，但这种关系受到天气的混淆，例如，天气越暖，冰淇淋的销量和犯罪率都会增加。
（当然，这过于简单化了，因为天气本身并不会导致犯罪，但它位于因果路径上。）

考虑一个思想实验，你身处一个黑暗的房间。
你的目标是预测犯罪，但你不知道天气或一年中的时间。
但是，你确实有关于冰淇淋销售的信息。
从因果角度来看，一个包含冰淇淋销量对犯罪影响的模型是有偏的——即使该模型显示出某种影响，冰淇淋销量也不会导致犯罪——但它会为你的犯罪模型提供一些预测价值。
这两种情况的原因相同：天气和冰淇淋销量是相关的，天气和犯罪也是相关的。
冰淇淋销量可以成功地（即使不完美地）充当天气的替代指标。
这导致了冰淇淋销量对犯罪的因果影响的有偏效应估计，但对犯罪的预测却部分有效。
其他变量也是如此，这些变量从因果角度来看是无效的，无论是由于自身有偏，还是由于将偏差引入因果效应估计中，通常都会带来良好的预测价值。
因此，预测准确性不是衡量因果关系的好方法。
