## 对于随机实验

由于$Y(1), Y(0) \perp D$，所以

$$
\widehat{A T E}=E_n[Y \mid D=1]-E_n[Y \mid D=0]
$$

因此

$$
\begin{aligned}
\widehat{A T E} & =E_n[Y \mid D=1]-E_n[Y \mid D=0] \\
& =\frac{1}{n_1} \sum_{i: D_i=1} Y_i-\frac{1}{n_0} \sum_{i: D_i=0} Y_i \\
& =\frac{1}{n_1} \sum_{i=1}^n D_i Y_i-\frac{1}{n_0} \sum_{i=1}^n\left(1-D_i\right) Y_i \\
& =\frac{\sum_{i=1}^n D_i Y_i}{\sum_{j=1}^n D_j}-\frac{\sum_{i=1}^n\left(1-D_i\right) Y_i}{\sum_{j=1}^n\left(1-D_j\right)} \\
& =\frac{n^{-1} \sum_{i=1}^n D_i Y_i}{n^{-1} \sum_{j=1}^n D_j}-\frac{n^{-1} \sum_{i=1}^n\left(1-D_i\right) Y_i}{n^{-1} \sum_{j=1}^n\left(1-D_j\right)} \\
& =\frac{1}{n} \sum_{i=1}^n \frac{D_i}{\hat{p}} Y_i-\frac{1}{n} \sum_{i=1}^n \frac{\left(1-D_i\right)}{1-\hat{p}} Y_i \\
& =E_n\left[\frac{D_i}{\hat{p}} Y_i\right]-E_n\left[\frac{\left(1-D_i\right)}{1-\hat{p}} Y_i\right]
\end{aligned}
$$

其中$\hat{p}=n^{-1} \sum_{i=1}^n D_i$
