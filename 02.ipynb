## 因果推断和理想实验

## 主要内容

## 政策评估问题

政策/项目评估的目标是评估政策干预的因果效应。例如： -
职业培训项目对收入和就业的影响 - 班级规模对考试成绩的影响 -
最低工资对就业的影响 - 建造地铁对于房价的影响

## 潜在结果的因果关系

### 处理

$D_i$: 单位 $i$ 接受处理的指标

$D_i = \begin{cases} 1 & \text{如果单位 } i \text{ 接受了处理} \\ 0 & \text{否则。} \end{cases}$

结果 - $Y_i$: 单位 $i$ 感兴趣的观察结果变量

潜在结果 - $Y_{0i}$ 和 $Y_{1i}$: 单位 $i$ 的潜在结果

于是

- $Y_{1i}$: 单位 $i$ 接受处理的潜在结果

- $Y_{0i}$: 单位 $i$ 未接受处理的潜在结果

### 处理效应

处理效应，或处理对单元 i 的结果的因果效应，是其两个潜在结果之间的差异：

$Y_{1i} - Y_{0i}$

观察到的结果

观察到的结果实现为

$Y_i = Y_{1i}D_i + Y_{0i}(1 - D_i)$ 或
$Y_i = \begin{cases} Y_{1i} & \text{如果 } D_i = 1 \\ Y_{0i} & \text{如果 } D_i = 0 \end{cases}$

因果推断的基本问题

无法观察到两个潜在结果 ($Y_{1i}$, $Y_{0i}$)

## 稳定单元处理值假设 (SUTVA)

假设观察到的结果实现为

$$Y_i = Y_{1i}D_i + Y_{0i}(1 - D_i)$$

- 意味着单元 i 的潜在结果不受单元 j 的处理的影响
- 排除单元间的干扰
- 例子：流感疫苗对住院的影响
- 这个假设可能存在问题，因此我们应该选择分析单元以最大限度地减少单元间的干扰。

## 关注的估计量

**ATE**

平均处理效应是：

$\alpha_{ATE} = E[Y_1 - Y_0]$

**ATET**

对已处理者的平均处理效应是：

$\alpha_{ATET} = E[Y_1 - Y_0|D = 1]$

**平均处理效应 (ATE)**

想象一个包含 4 个单元的总体：

| i | $Y_{1i}$ （如果接受处理的潜在结果） | $Y_{0i}$ （如果未接受处理的潜在结果） | $Y_i$ （实际结果） | $D_i$ （处理状态） | $Y_{1i} - Y_{0i}$ （个体处理效应） |
|------|-------------|-------------|---------|---------|------------------------|
| 1 | 3 | 0 | 3 | 1 | 3 |
| 2 | 1 | 1 | 1 | 1 | 0 |
| 3 | 1 | 0 | 0 | 0 | 1 |
| 4 | 1 | 1 | 1 | 0 | 0 |
| $E[Y_1]$ （接受处理的期望结果） | 1.5 |  |  |  |  |
| $E[Y_0]$ （未接受处理的期望结果） |  | 0.5 |  |  |  |
| $E[Y_1 - Y_0]$ （平均个体处理效应的期望） |  |  |  | 1 |  |

<br>

$$\alpha_{ATE} = E[Y_1 - Y_0] = 3 \cdot (1/4) + 0 \cdot (1/4) + 1 \cdot (1/4) + 0 \cdot (1/4) = 1$$
（平均处理效应）

**平均处理效应对受处理者的影响 (ATET)**

想象一个有 4 个单位的总体：

| i | $Y_{1i}$ （如果接受处理的结果） | $Y_{0i}$ （如果未接受处理的结果） | $Y_i$ （实际结果） | $D_i$ （是否接受处理，1=是，0=否） | $Y_{1i} - Y_{0i}$ （处理效应） |
|-----|--------------|--------------|---------|---------|------------------------|
| 1 | 3 | 0 | 3 | 1 | 3 |
| 2 | 1 | 1 | 1 | 1 | 0 |
| 3 | 1 | 0 | 0 | 0 | 1 |
| 4 | 1 | 1 | 1 | 0 | 0 |
|  |  |  |  |  |  |
| $E[Y_1|D=1]$ （当D=1时，$Y_1$的期望） | 2 |  |  |  |  |
| $E[Y_0|D=1]$ （当D=1时，$Y_0$的期望） |  | 0.5 |  |  |  |
| $E[Y_1 - Y_0|D=1]$ （当D=1时，$Y_1 - Y_0$的期望） |  |  |  |  | 1.5 |

<br>

$$\alpha_{ATET} = E[Y_1 - Y_0 | D=1] = 3 \cdot (1/2) + 0 \cdot (1/2) = 1.5$$

## 选择偏差

### 问题

对已处理和未处理对象的收入进行比较通常不会给出正确的答案：

$$
\begin{array}{r}
E[Y \mid D=1]-E[Y \mid D=0]=E\left[Y_1 \mid D=1\right]-E\left[Y_0 \mid D=0\right] \\
=\underbrace{E\left[Y_1-Y_0 \mid D=1\right]}_{A T E T}+\underbrace{\left\{E\left[Y_0 \mid D=1\right]-E\left[Y_0 \mid D=0\right]\right\}}_{B I A S}
\end{array}
$$

示例：针对弱势群体的就业培训项目

- 参与者是从劳动状况困难的个体亚群体中自我选择的

- 在没有该项目的情况下，参与者在培训后的收入将低于非参与者

  ($E[Y_0|D = 1] - E[Y_0|D = 0] < 0$)

## 关键思想

- 因果关系由潜在结果定义，而不是由已实现的（观察到的）结果定义。
- 观察到的关联对于因果关系来说既不是必要的也不是充分的。
- 估计处理的因果效应（通常）从研究分配机制开始。

## 随机分配解决选择性问题

回顾在比较已处理组和未处理组的平均结果时遇到的选择问题：

$$
\begin{aligned}
& \underbrace{E[Y \mid D=1]-E[Y \mid D=0]}_{\text {Difference in Means }}=E\left[Y_1 \mid D=1\right]-E\left[Y_0 \mid D=0\right] \\
& \quad=\underbrace{E\left[Y_1-Y_0 \mid D=1\right]}_{\text {ATET }}+\underbrace{\left\{E\left[Y_0 \mid D=1\right]-E\left[Y_0 \mid D=0\right]\right\}}_{\text {BIAS }}
\end{aligned}
$$

偏差（BIAS）

- 将单位随机分配到处理组会迫使选择偏差为零
- 处理组和对照组在所有特征（包括$Y_0$）上都趋于相似

随机化意味着：

$(Y_1, Y_0)$ 独立于 $D$，或者 $\left(Y_1, Y_0\right) \perp D$ 。

我们有 $E[Y_0|D = 1] = E[Y_0|D = 0]$，因此

$$\alpha_{ATET} = E[Y_1 - Y_0|D = 1] = E[Y|D = 1] - E[Y|D = 0]$$

此外，我们有

$$\alpha_{ATE} = E[Y_1 - Y_0] = E[Y_1 - Y_0|D = 1] = E[Y|D = 1] - E[Y|D = 0]$$

因此，

$$\underbrace{E[Y|D = 1] - E[Y|D = 0]}_{\text{均值差异}} = \alpha_{ATE} = \alpha_{ATET}$$

### 随机实验中的识别

识别结果超越了平均处理效应。

令 $Q_\theta(Y)$ 为 $Y$ 分布的第 $\theta$ 个分位数：

$$Pr(Y \leq Q_\theta(Y)) = \theta$$.

给定随机分配，$Y_0 \perp D$。因此，

$$Y_0 \sim Y_0|D = 0 \sim Y|D = 0$$

其中“$\sim$”表示“具有相同的分布”。类似地，

$$Y_1 \sim Y|D = 1$$.

因此，在任何分位数上，处理效应 $Q_\theta(Y_1) - Q_\theta(Y_0)$
都是可以识别的。

- 随机化识别了 $Y_0$ 和 $Y_1$ 的整个边际分布。
- 无法识别效应的分位数：$Q_\theta(Y_1 - Y_0)$（分位数的差不是差的分位数）。

### 随机试验中的估计

考虑一个有 N 个个体的随机试验。假设感兴趣的估计量是 ATE：

$$\alpha_{ATE} = E[Y_1 - Y_0] = E[Y|D = 1] - E[Y|D = 0].$$

使用类比原理，我们构建一个估计器：

$$\hat{\alpha} = \bar{Y}_1 - \bar{Y}_0,$$

其中

$$\bar{Y}_1 = \frac{\sum Y_i \cdot D_i}{\sum D_i} = \frac{1}{N_1} \sum_{D_i=1} Y_i;$$

$$\bar{Y}_0 = \frac{\sum Y_i \cdot (1 - D_i)}{\sum (1 - D_i)} = \frac{1}{N_0} \sum_{D_i=0} Y_i$$

其中 $N_1 = \sum_i D_i$ 且 $N_0 = N - N_1$.

$\hat{\alpha}$ 是 $\alpha_{ATE}$ 的无偏且一致的估计量。

## 协变量平衡

- 随机化平衡了观察到的和未观察到的特征，使得处理组（treatment）和对照组之间具有可比性。

- 可以使用所谓的“平衡测试”（例如，t检验）来检查随机分配，以查看观察到的协变量$X$的分布在处理组和对照组中是否相同。

- $X$是处理前变量，在处理分配之前测量（即，在“基线”时）。

## 随机实验有效性的威胁

- 内部有效性：我们能否评估我们特定样本的治疗效果？
  - 当处理组和对照组之间存在差异（除了处理本身），这些差异会影响结果，且我们无法控制时，则失败。

<br>

- 外部有效性：我们能否将评估结果推广到其他人群？
  - 当评估效果在评估环境之外不同时，则失败。
